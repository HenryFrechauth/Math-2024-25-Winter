# Useful Formulas and Notes

## 1. Equation of a line passing through points $A(1, 2)$ and $B(3, 4)$
- **Formula for the slope $m$**:
  $$
  m = \frac{y_2 - y_1}{x_2 - x_1}
  $$
- **Point-slope form**:
  $$
  y - y_1 = m(x - x_1)
  $$

## 2. Equation of a line passing through point $A(1, 2)$ and parallel to $y = 2x + 3$
- A parallel line has the **same slope**. If $y = mx + b$, the slope is $m$.
- Use the **point-slope form**:
  $$
  y - y_1 = m(x - x_1)
  $$

## 3. Equation of a line passing through point $A(1, 2)$ and perpendicular to $y = 2x + 3$
- **Perpendicular slope**:
  $$
  m_{\text{perpendicular}} = -\frac{1}{m}
  $$
  where $m$ is the slope of the given line.
- Use the **point-slope form**:
  $$
  y - y_1 = m(x - x_1)
  $$

## 4. Intersection and angle between $y = 2x + 3$ and $y = 3x + 2$
- **Intersection point**: Solve the system of equations:
  $$
  y = 2x + 3, \quad y = 3x + 2
  $$
- **Angle between lines**:
  $$
  \tan \theta = \left|\frac{m_2 - m_1}{1 + m_1m_2}\right|
  $$
  where $m_1, m_2$ are the slopes of the lines.

## 5. Line passing through $A(1, 2)$ and parallel to vector $\vec{v} = [2, 3]$
- A direction vector $[a, b]$ gives the slope:
  $$
  m = \frac{b}{a}
  $$
- Use the **point-slope form**:
  $$
  y - y_1 = m(x - x_1)
  $$

## 6. Example of a line perpendicular and parallel to $y = 2x + 3$
- **Parallel**: Same slope $m = 2$, e.g., $y = 2x + c$, where $c$ is a constant.
- **Perpendicular**: Slope $m = -\frac{1}{2}$, e.g., $y = -\frac{1}{2}x + c$.

## 7. Distance from $A(1, 2)$ to the line $y = 2x + 3$
- Convert to standard form $Ax + By + C = 0$:
  $$
  -2x + y - 3 = 0
  $$
- **Distance formula**:
  $$
  d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
  $$

## 8. Line intersecting coordinate axes at $A(2, 0)$ and $B(0, 3)$
- Use the **two-point form**:
  $$
  \frac{x}{x_1} + \frac{y}{y_2} = 1
  $$
  Simplify to get the equation.

## 9. Angle between the line $y = x + 3$ and the $Ox$-axis
- Slope $m = 1$.
- **Angle with $Ox$-axis**:
  $$
  \theta = \arctan(m)
  $$

## 10. Perpendicular vector to $x + y + 1 = 0$
- In the standard form $Ax + By + C = 0$, a vector perpendicular to the line is:
  $$
  \vec{n} = [A, B]
  $$
  For $x + y + 1 = 0$, $\vec{n} = [1, 1]$.

To solve for the intersection point and angle between the lines
\( y = 2x + 3 \) and \( y = 3x + 2 \), follow these steps:

---

### 1. Find the Intersection Point

The intersection point is where the \( y \)-values of the two equations are equal. Set the equations equal to each other:
$$
2x + 3 = 3x + 2
$$

Solve for \( x \):
$$
3 - 2 = 3x - 2x
$$
$$
x = 1
$$

Substitute \( x = 1 \) into either equation to find \( y \). Using \( y = 2x + 3 \):
$$
y = 2(1) + 3 = 5
$$

Thus, the intersection point is:
$$
(1, 5)
$$

---

### 2. Calculate the Angle Between the Lines

The formula for the angle \( \theta \) between two lines is:
$$
\tan\theta = \left| \frac{m_2 - m_1}{1 + m_1m_2} \right|
$$
Here, \( m_1 = 2 \) (slope of \( y = 2x + 3 \)) and \( m_2 = 3 \) (slope of \( y = 3x + 2 \)).

Substitute the values:
$$
\tan\theta = \left| \frac{3 - 2}{1 + (2)(3)} \right|
$$
$$
\tan\theta = \left| \frac{1}{1 + 6} \right|
$$
$$
\tan\theta = \frac{1}{7}
$$

Now, find \( \theta \) using the arctangent function:
$$
\theta = \arctan\left(\frac{1}{7}\right)
$$

---

### Final Results

1. **Intersection Point**:
   $$
   (1, 5)
   $$

2. **Angle Between the Lines**:
   $$
   \theta = \arctan\left(\frac{1}{7}\right) \approx 8.13^\circ
   $$

# Problem: Finding Lines Perpendicular and Parallel to a Given Line

We are given the line:
$$
y = 2x + 3
$$

---

### 1. Parallel Line

A line parallel to the given line will have the same slope \( m = 2 \). Using the point-slope form, lets construct a parallel line passing through a different point, e.g., \( (0, 1) \).

The equation is:
$$
y - y_1 = m(x - x_1)
$$
Substitute \( m = 2 \), \( x_1 = 0 \), and \( y_1 = 1 \):
$$
y - 1 = 2(x - 0)
$$
Simplify:
$$
y = 2x + 1
$$

Thus, an example of a parallel line is:
$$
y = 2x + 1
$$

---

### 2. Perpendicular Line

A line perpendicular to the given line will have a slope that is the negative reciprocal of \( m = 2 \). Thus, the slope of the perpendicular line is:
$$
m = -\frac{1}{2}
$$

Using the point-slope form, lets construct a perpendicular line passing through \( (0, 1) \).

The equation is:
$$
y - y_1 = m(x - x_1)
$$
Substitute \( m = -\frac{1}{2} \), \( x_1 = 0 \), and \( y_1 = 1 \):
$$
y - 1 = -\frac{1}{2}(x - 0)
$$
Simplify:
$$
y = -\frac{1}{2}x + 1
$$

Thus, an example of a perpendicular line is:
$$
y = -\frac{1}{2}x + 1
$$

---

### Final Results

1. **Example of a Parallel Line**:
   $$
   y = 2x + 1
   $$

2. **Example of a Perpendicular Line**:
   $$
   y = -\frac{1}{2}x + 1
   $$

# Problem: Distance from a Point to a Line

We are given the line:
$$
y = 2x + 3
$$
and the point:
$$
A(1, 2)
$$

---

### Formula for Distance from a Point to a Line

The general equation of a line is:
$$
Ax + By + C = 0
$$
For the given line \( y = 2x + 3 \), rewrite it in general form:
$$
2x - y + 3 = 0
$$
Here, \( A = 2 \), \( B = -1 \), and \( C = 3 \).

The distance \( d \) from a point \( (x_1, y_1) \) to the line \( Ax + By + C = 0 \) is given by:
$$
d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$

---

### Substitute Values into the Formula

For the point \( A(1, 2) \):
- \( x_1 = 1 \)
- \( y_1 = 2 \)
- \( A = 2 \)
- \( B = -1 \)
- \( C = 3 \)

Substitute into the formula:
$$
d = \frac{|2(1) + (-1)(2) + 3|}{\sqrt{2^2 + (-1)^2}}
$$
Simplify the numerator:
$$
d = \frac{|2 - 2 + 3|}{\sqrt{4 + 1}}
$$
$$
d = \frac{|3|}{\sqrt{5}}
$$
$$
d = \frac{3}{\sqrt{5}}
$$

---

### Final Answer

The distance from \( A(1, 2) \) to the line \( y = 2x + 3 \) is:
$$
d = \frac{3}{\sqrt{5}}
$$

For a rationalized denominator:
$$
d = \frac{3\sqrt{5}}{5}
$$

# Problem: Find the Equation of the Line

The line intersects the coordinate axes at points:
$$
A(2, 0) \quad \text{and} \quad B(0, 3)
$$

---

### 1. Use the Two-Point Form of a Line

The two-point form of the equation of a line is:
$$
y - y_1 = \frac{y_2 - y_1}{x_2 - x_1}(x - x_1)
$$
Here, \( A(2, 0) \) and \( B(0, 3) \). Let:
- \( (x_1, y_1) = (2, 0) \)
- \( (x_2, y_2) = (0, 3) \)

---

### 2. Substitute the Values

Substitute into the formula:
$$
y - 0 = \frac{3 - 0}{0 - 2}(x - 2)
$$
Simplify the slope:
$$
\text{slope} = \frac{3}{-2} = -\frac{3}{2}
$$
So the equation becomes:
$$
y = -\frac{3}{2}(x - 2)
$$

---

### 3. Simplify the Equation

Expand and simplify:
$$
y = -\frac{3}{2}x + 3
$$

---

### Final Equation

The equation of the line is:
$$
y = -\frac{3}{2}x + 3
$$

$$
y = x + 3
$$

---

### 1. Slope of the Line

$$

---

### 2. Formula for the Angle Between a Line and the \( O_x \)-Axis

$$
\theta = \arctan(m)

# Problem: Find a Vector Perpendicular to the Line

We are given the line in standard form:
$$
x + y + 1 = 0
$$

---

### 1. General Representation of a Line

In the general form of a line:
$$
Ax + By + C = 0
$$
The vector \( \vec{n} = [A, B] \) is normal (perpendicular) to the line.

For the given line:
$$
A = 1, \quad B = 1, \quad C = 1
$$

Thus, the normal vector is:
$$
\vec{n} = [1, 1]
$$

---

### 2. Verify Perpendicularity

The direction vector of the line \( \vec{v} \), which is parallel to the line, is:
$$
\vec{v} = [-B, A] = [-1, 1]
$$

The dot product of \( \vec{n} \) and \( \vec{v} \) confirms perpendicularity:
$$
\vec{n} \cdot \vec{v} = (1)(-1) + (1)(1) = -1 + 1 = 0
$$

Since the dot product is zero, \( \vec{n} \) is perpendicular to the line.

---

### Final Answer

A vector perpendicular to the line \( x + y + 1 = 0 \) is:
$$
\vec{n} = [1, 1]
$$
