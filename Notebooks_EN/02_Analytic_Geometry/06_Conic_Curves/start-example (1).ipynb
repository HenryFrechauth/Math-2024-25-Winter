# Conic Sections - Useful Formulas and Notes

## 1. Equation of a circle with center at $A(1, 2)$ and radius $r=3$
- **Standard form of a circle**:
  $$
  (x - h)^2 + (y - k)^2 = r^2
  $$
  where $(h, k)$ is the center and $r$ is the radius.

## 2. Equation of a parabola intersecting the $Ox$ axis at points $x=2$, $x=4$, and passing through $y(3)=1$
- A parabola intersecting the $Ox$ axis at $x=a$ and $x=b$ can be written as:
  $$
  y = k(x - a)(x - b)
  $$
- Use the given point $(3, 1)$ to find $k$ by substitution.

## 3. Center of the ellipse with the equation $x^2 + 4y^2 - 4x - 16y + 16 = 0$
- Rewrite the equation in standard form by **completing the square**:
  1. Group $x$-terms and $y$-terms.
  2. Complete the square for each variable.
- Standard form of an ellipse:
  $$
  \frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
  $$
  where $(h, k)$ is the center.

## 4. Slope ($m > 0$) of the line $y=mx-5$ tangent to the circle $x^2 + y^2 = 1$
- The condition for tangency between a line $y = mx + c$ and a circle $x^2 + y^2 = r^2$:
  $$
  \frac{|c|}{\sqrt{1 + m^2}} = r
  $$
- For $r = 1$ and $c = -5$, solve for $m > 0$.

## 5. Intersection points of the hyperbola $x^2 - y^2 = 1$ with the ellipse $x^2 + 4y^2 = 6$
- Solve the system of equations:
  1. $x^2 - y^2 = 1$
  2. $x^2 + 4y^2 = 6$
- Use substitution or elimination to find the points of intersection.

## 6. Distance between the branches of the hyperbola $x^2 - y^2 = 1$
- **Standard form of a hyperbola**:
  $$
  \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
  $$
- The distance between the branches is:
  $$
  \text{Distance} = 2b
  $$
  where $b = \sqrt{c^2 - a^2}$ and $c$ is the distance from the center to the foci.

$$
(x - h)^2 + (y - k)^2 = r^2

Where:
- \((h, k)\) represents the coordinates of the center of the circle.
- \(r\) is the radius of the circle.

---

$$
(x - 1)^2 + (y - 2)^2 = 3^2
$$

---

$$
r^2 = 3^2 = 9
$$

Substitute this value into the equation:

$$
(x - 1)^2 + (y - 2)^2 = 9
$$

---

### Step 4: Final Equation of the Circle
The simplified equation of the circle is:

$$
(x - 1)^2 + (y - 2)^2 = 9
$$

---

### Summary

$$
(x - 1)^2 + (y - 2)^2 = 9

### Step 1: General Form of a Parabola
A parabola that intersects the \(x\)-axis at two points has the general factored form:

$$
y = a(x - x_1)(x - x_2)
$$

Where:
- \(x_1\) and \(x_2\) are the roots (x-intercepts) of the parabola.
- \(a\) is a constant that determines the vertical stretch or compression of the parabola.

---

### Step 2: Substituting the Given Roots
The parabola intersects the \(x\)-axis at \(x = 2\) and \(x = 4\). Thus, the equation becomes:

$$
y = a(x - 2)(x - 4)
$$

---

### Step 3: Using the Given Point to Find \(a\)
The parabola passes through the point \((3, 1)\). Substituting \(x = 3\) and \(y = 1\) into the equation:

$$
1 = a(3 - 2)(3 - 4)
$$

Simplify the terms inside the parentheses:

$$
1 = a(1)(-1)
$$

This reduces to:

$$
1 = -a
$$

Solve for \(a\):

$$
a = -1
$$

---

### Step 4: Final Equation of the Parabola
Substitute \(a = -1\) back into the equation:

$$
y = -1(x - 2)(x - 4)
$$

Simplify the equation by expanding:

$$
y = -(x^2 - 6x + 8)
$$

Distribute the negative sign:

$$
y = -x^2 + 6x - 8
$$

---

### Summary
The equation of the parabola that intersects the \(x\)-axis at \(x = 2\) and \(x = 4\), and passes through the point \((3, 1)\), is:

$$
y = -x^2 + 6x - 8
$$

### Step 1: General Form of the Ellipse Equation
The given equation of the ellipse is:

$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$

To find the center, we need to rewrite this equation in the standard form of an ellipse:

$$
\frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
$$

This involves completing the square for both \(x\) and \(y\).

---

### Step 2: Rearrange the Terms
Group the \(x\)-terms and \(y\)-terms together:

$$
(x^2 - 4x) + (4y^2 - 16y) = -16
$$

Factor out \(4\) from the \(y^2\)-terms:

$$
(x^2 - 4x) + 4(y^2 - 4y) = -16
$$

---

### Step 3: Complete the Square
#### For the \(x\)-terms:
Take half the coefficient of \(x\), square it, and add it inside the parentheses:

- Half of \(-4\) is \(-2\), and \((-2)^2 = 4\).

Add \(4\) to \(x^2 - 4x\):

$$
(x^2 - 4x + 4)
$$

#### For the \(y\)-terms:
Take half the coefficient of \(y\), square it, and add it inside the parentheses:

- Half of \(-4\) is \(-2\), and \((-2)^2 = 4\).

Add \(4\) to \(y^2 - 4y\):

$$
(y^2 - 4y + 4)
$$

But since the \(y^2\)-terms are multiplied by \(4\), adding \(4\) inside the parentheses actually adds \(16\) to the equation.

---

### Step 4: Update the Equation
The equation becomes:

$$
(x^2 - 4x + 4) + 4(y^2 - 4y + 4) = -16 + 4 + 16
$$

Simplify:

$$
(x - 2)^2 + 4(y - 2)^2 = 4
$$

---

### Step 5: Standard Form of the Ellipse
Divide the entire equation by \(4\) to write it in standard form:

$$
\frac{(x - 2)^2}{4} + \frac{(y - 2)^2}{1} = 1
$$

---

### Step 6: Identify the Center
The center of the ellipse is at:

$$
(h, k) = (2, 2)
$$

---

### Summary
The center of the ellipse with the given equation is:

$$
(2, 2)
$$

### Step 1: Equations of the Line and the Circle
We are given:
1. The equation of the line:
   $$
   y = mx - 5
   $$
2. The equation of the circle:
   $$
   x^2 + y^2 = 1
   $$

The goal is to find the value of \(m > 0\) such that the line is tangent to the circle.

---

### Step 2: Substituting the Line Equation into the Circle Equation
Substitute \(y = mx - 5\) into the circle equation \(x^2 + y^2 = 1\):

$$
x^2 + (mx - 5)^2 = 1
$$

Expand \((mx - 5)^2\):

$$
x^2 + (m^2x^2 - 10mx + 25) = 1
$$

Simplify:

$$
x^2 + m^2x^2 - 10mx + 25 = 1
$$

Combine like terms:

$$
(1 + m^2)x^2 - 10mx + 24 = 0
$$

---

### Step 3: Condition for Tangency
For the line to be tangent to the circle, the quadratic equation in \(x\) must have exactly one solution. This happens when the discriminant \(\Delta\) of the quadratic equation is zero.

The discriminant of a quadratic equation \(ax^2 + bx + c = 0\) is given by:

$$
\Delta = b^2 - 4ac
$$

In our equation:

- \(a = 1 + m^2\)
- \(b = -10m\)
- \(c = 24\)

Substitute into the discriminant formula:

$$
\Delta = (-10m)^2 - 4(1 + m^2)(24)
$$

Simplify:

$$
\Delta = 100m^2 - 4(1 + m^2)(24)
$$

Expand the second term:

$$
\Delta = 100m^2 - 96 - 96m^2
$$

Simplify:

$$
\Delta = 4m^2 - 96
$$

---

### Step 4: Solve for \(m\) When \(\Delta = 0\)
Set the discriminant to zero:

$$
4m^2 - 96 = 0
$$

Solve for \(m^2\):

$$
4m^2 = 96
$$

$$
m^2 = 24
$$

Take the square root of both sides:

$$
m = \sqrt{24} = 2\sqrt{6}
$$

Since \(m > 0\), we have:

$$
m = 2\sqrt{6}
$$

---

### Summary
The slope of the line \(y = mx - 5\) that is tangent to the circle \(x^2 + y^2 = 1\) is:

$$
m = 2\sqrt{6}
$$

### Step 1: Equations of the Hyperbola and Ellipse
We are given:
1. The equation of the hyperbola:
   $$
   x^2 - y^2 = 1
   $$
2. The equation of the ellipse:
   $$
   x^2 + 4y^2 = 6
   $$

The goal is to find the points of intersection of these two curves.

---

### Step 2: Solving the System of Equations
We solve the system of equations:
1. \(x^2 - y^2 = 1\)
2. \(x^2 + 4y^2 = 6\)

#### Step 2.1: Eliminate \(x^2\)
From the first equation, solve for \(x^2\):

$$
x^2 = y^2 + 1
$$

Substitute this expression for \(x^2\) into the second equation:

$$
(y^2 + 1) + 4y^2 = 6
$$

Simplify:

$$
y^2 + 1 + 4y^2 = 6
$$

Combine like terms:

$$
5y^2 + 1 = 6
$$

Subtract \(1\) from both sides:

$$
5y^2 = 5
$$

Solve for \(y^2\):

$$
y^2 = 1
$$

#### Step 2.2: Solve for \(y\)
Take the square root of both sides:

$$
y = \pm 1
$$

---

### Step 3: Solve for \(x\)
Substitute \(y = 1\) and \(y = -1\) into the expression \(x^2 = y^2 + 1\):

For \(y = 1\):

$$
x^2 = 1^2 + 1 = 2
$$

Take the square root:

$$
x = \pm\sqrt{2}
$$

For \(y = -1\):

$$
x^2 = (-1)^2 + 1 = 2
$$

Take the square root:

$$
x = \pm\sqrt{2}
$$

---

### Step 4: Intersection Points
The intersection points are:

$$
1. (\sqrt{2}, 1)
2. (- \sqrt{2}, 1)
3. (\sqrt{2}, -1)
4. (- \sqrt{2}, -1)
$$
---

### Summary
The hyperbola \(x^2 - y^2 = 1\) and the ellipse \(x^2 + 4y^2 = 6\) intersect at the points:

$$
(\sqrt{2}, 1), (-\sqrt{2}, 1), (\sqrt{2}, -1), (-\sqrt{2}, -1)
$$

### Step 1: General Form of the Hyperbola
The given hyperbola is:

$$
x^2 - y^2 = 1
$$

This is a hyperbola centered at the origin \((0, 0)\) and oriented horizontally. The general equation of a hyperbola in this form is:

$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
$$

Here, \(a^2 = 1\) and \(b^2 = 1\), so:

- \(a = 1\): the distance from the center to the vertices along the \(x\)-axis.
- \(b = 1\): the value determining the slope of the asymptotes.

---

### Step 2: Vertical Distance Between Branches
The two branches of the hyperbola are symmetric about the \(x\)-axis, and the vertical distance between them at a given \(x\)-coordinate is determined by solving for \(y\) in terms of \(x\).

From the equation \(x^2 - y^2 = 1\), solve for \(y^2\):

$$
y^2 = x^2 - 1
$$

Take the square root to find \(y\):

$$
y = \pm \sqrt{x^2 - 1}
$$

The vertical distance between the branches is:

$$
\text{Distance} = y_{\text{top}} - y_{\text{bottom}} = 2\sqrt{x^2 - 1}
$$

---

### Step 3: Smallest Distance Between Branches
The branches are closest to each other near the vertices of the hyperbola, where \(x = \pm 1\). At these points:

$$
y = \pm \sqrt{1^2 - 1} = \pm \sqrt{0} = 0
$$

At \(x = \pm 1\), the branches touch the \(x\)-axis, so the **distance between the branches is zero at this point**.

---

### Step 4: Distance Away from the Vertices
As \(x\) increases in magnitude (\(|x| > 1\)), the distance between the branches grows, following the formula:

$$
\text{Distance} = 2\sqrt{x^2 - 1}
$$

---

### Summary
For the hyperbola \(x^2 - y^2 = 1\):
- The **distance between the branches** at a given \(x > 1\) is:

$$
\text{Distance} = 2\sqrt{x^2 - 1}
$$

- The branches are **closest at the vertices** (\(x = \pm 1\)), where the distance is zero.
