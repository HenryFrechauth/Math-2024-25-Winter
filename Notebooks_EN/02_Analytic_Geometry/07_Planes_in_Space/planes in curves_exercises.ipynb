To find the equation of the plane passing through the points \( A(1, 2, 3) \), \( B(3, 4, 5) \), and \( C(2, 1, 4) \), follow these steps:

### Step 1: Find two vectors lying on the plane.
We can construct vectors \( \mathbf{AB} \) and \( \mathbf{AC} \) using the coordinates of the points:
$$
\mathbf{AB} = B - A = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2),
$$
$$
\mathbf{AC} = C - A = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1).
$$

### Step 2: Compute the normal vector to the plane.
The normal vector \( \mathbf{n} \) to the plane can be found using the cross product of \( \mathbf{AB} \) and \( \mathbf{AC} \):
$$
\mathbf{n} = \mathbf{AB} \times \mathbf{AC}.
$$
$$
\mathbf{n} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
2 & 2 & 2 \\
1 & -1 & 1
\end{vmatrix}.
$$
Expand the determinant:
$$
\mathbf{n} = \mathbf{i} \begin{vmatrix} 2 & 2 \\ -1 & 1 \end{vmatrix}
- \mathbf{j} \begin{vmatrix} 2 & 2 \\ 1 & 1 \end{vmatrix}
+ \mathbf{k} \begin{vmatrix} 2 & 2 \\ 1 & -1 \end{vmatrix}.
$$
$$
\mathbf{n} = \mathbf{i} (2 \cdot 1 - 2 \cdot -1)
- \mathbf{j} (2 \cdot 1 - 2 \cdot 1)
+ \mathbf{k} (2 \cdot -1 - 2 \cdot 1).
$$
$$
\mathbf{n} = \mathbf{i} (2 + 2) - \mathbf{j} (2 - 2) + \mathbf{k} (-2 - 2).
$$
$$
\mathbf{n} = 4\mathbf{i} - 0\mathbf{j} - 4\mathbf{k}.
$$
$$
\mathbf{n} = (4, 0, -4).
$$

### Step 3: Write the equation of the plane.
The general equation of a plane is:
$$
ax + by + cz + d = 0,
$$
where \( (a, b, c) \) is the normal vector. Using \( \mathbf{n} = (4, 0, -4) \), the equation becomes:
$$
4x + 0y - 4z + d = 0,
$$
or:
$$
4x - 4z + d = 0.
$$

### Step 4: Find \( d \) using a point on the plane.
Substitute \( A(1, 2, 3) \) into the plane equation:
$$
4(1) - 4(3) + d = 0.
$$
$$
4 - 12 + d = 0.
$$
$$
d = 8.
$$

### Final Equation of the Plane:
$$
4x - 4z + 8 = 0,
$$
or equivalently:
$$
x - z + 2 = 0.
$$

4.

To find the line of intersection of the planes \( 2x + 3y + 4z = 5 \) and \( 3x + 4y + 2z = 6 \), follow these steps:

---

### Step 1: Represent the system of equations
The equations of the planes are:
1. \( 2x + 3y + 4z = 5 \)
2. \( 3x + 4y + 2z = 6 \)

The line of intersection will satisfy both equations. This is a system of linear equations in \( x \), \( y \), and \( z \).

---

### Step 2: Eliminate one variable
We eliminate \( z \) by multiplying the equations to align their coefficients for \( z \):
- Multiply the first equation by 2:
$$
4x + 6y + 8z = 10.
$$
- Multiply the second equation by 4:
$$
12x + 16y + 8z = 24.
$$

Subtract the first equation from the second:
$$
(12x + 16y + 8z) - (4x + 6y + 8z) = 24 - 10.
$$
$$
8x + 10y = 14.
$$
Simplify:
$$
4x + 5y = 7. \tag{3}
$$

---

### Step 3: Parametrize the solution
Solve equation (3) for \( x \) in terms of \( y \):
$$
x = \frac{7 - 5y}{4}. \tag{4}
$$

Substitute \( x \) from equation (4) into one of the original plane equations, say \( 2x + 3y + 4z = 5 \):
$$
2\left(\frac{7 - 5y}{4}\right) + 3y + 4z = 5.
$$
Simplify:
$$
\frac{14 - 10y}{4} + 3y + 4z = 5.
$$
Multiply through by 4 to eliminate fractions:
$$
14 - 10y + 12y + 16z = 20.
$$
$$
2y + 16z = 6.
$$
Simplify:
$$
y + 8z = 3. \tag{5}
$$

Solve equation (5) for \( y \) in terms of \( z \):
$$
y = 3 - 8z. \tag{6}
$$

---

### Step 4: Write the parametric equations of the line
Substitute \( y = 3 - 8z \) into \( x = \frac{7 - 5y}{4} \):
$$
x = \frac{7 - 5(3 - 8z)}{4}.
$$
Simplify:
$$
x = \frac{7 - 15 + 40z}{4}.
$$
$$
x = \frac{-8 + 40z}{4}.
$$
$$
x = -2 + 10z. \tag{7}
$$

From equations (6) and (7), the parametric equations of the line are:
$$
x = -2 + 10t, \quad y = 3 - 8t, \quad z = t,
$$
where \( t \) is a parameter.

---

### Final Answer:
The line of intersection is given by:
$$
\begin{aligned}
x &= -2 + 10t, \\
y &= 3 - 8t, \\
z &= t,
\end{aligned}
$$

To find the equation of the plane passing through the point \( A(1, 2, 3) \) and parallel to the vectors \( \mathbf{v}_1 = [1, 0, 1] \) and \( \mathbf{v}_2 = [0, 1, -1] \), follow these steps:

---

### Step 1: Normal vector to the plane

$
The plane is parallel to \( \mathbf{v}_1 \) and \( \mathbf{v}_2 \). The normal vector \( \mathbf{n} \) to the plane is perpendicular to both \( \mathbf{v}_1 \) and \( \mathbf{v}_2 \). We compute \( \mathbf{n} \) using the cross product:
$$
\mathbf{n} = \mathbf{v}_1 \times \mathbf{v}_2.
$$

Write \( \mathbf{v}_1 \) and \( \mathbf{v}_2 \) in component form:
$$
\mathbf{v}_1 = [1, 0, 1], \quad \mathbf{v}_2 = [0, 1, -1].
$$

Compute the cross product:
$$
\mathbf{n} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 0 & 1 \\
0 & 1 & -1
\end{vmatrix}.
$$

Expand the determinant:
$$
\mathbf{n} = \mathbf{i} \begin{vmatrix} 0 & 1 \\ 1 & -1 \end{vmatrix}
- \mathbf{j} \begin{vmatrix} 1 & 1 \\ 0 & -1 \end{vmatrix}
+ \mathbf{k} \begin{vmatrix} 1 & 0 \\ 0 & 1 \end{vmatrix}.
$$

Simplify each term:
$$
\mathbf{n} = \mathbf{i}(0 \cdot -1 - 1 \cdot 1) - \mathbf{j}(1 \cdot -1 - 1 \cdot 0) + \mathbf{k}(1 \cdot 1 - 0 \cdot 0).
$$
$$
\mathbf{n} = \mathbf{i}(-1) - \mathbf{j}(-1) + \mathbf{k}(1).
$$
$$
\mathbf{n} = -\mathbf{i} + \mathbf{j} + \mathbf{k}.
$$

The normal vector is:
$$
\mathbf{n} = [-1, 1, 1].
$$

---

### Step 2: Equation of the plane
The general equation of a plane is:
$$
n_1(x - x_0) + n_2(y - y_0) + n_3(z - z_0) = 0,
$$
where \( \mathbf{n} = [n_1, n_2, n_3] \) is the normal vector and \( (x_0, y_0, z_0) \) is a point on the plane.

Substitute \( \mathbf{n} = [-1, 1, 1] \) and \( A(1, 2, 3) \):
$$
-1(x - 1) + 1(y - 2) + 1(z - 3) = 0.
$$

Simplify:
$$
-(x - 1) + (y - 2) + (z - 3) = 0.
$$
$$
-x + 1 + y - 2 + z - 3 = 0.
$$
$$
-x + y + z - 4 = 0.
$$

---

### Final Equation of the Plane
The equation of the plane is:
$$
x - y - z + 4 = 0.
$$

For two planes to be parallel, they must have the same normal vector. The normal vector of the given plane is \( \mathbf{n} = [2, 3, 4] \), as it corresponds to the coefficients of \( x \), \( y \), and \( z \).

Thus, any plane parallel to the given plane will have the same normal vector. The equation of such a plane will be of the form:
$$
2x + 3y + 4z = D,

$$
2x + 3y + 4z = 10.
$$

---

### Step 2: Plane perpendicular to the given plane
For two planes to be perpendicular, their normal vectors must be perpendicular. The normal vector of the given plane is \( \mathbf{n} = [2, 3, 4] \).

To find a plane perpendicular to this one, we need to choose a normal vector that is perpendicular to \( \mathbf{n} \). The dot product of two perpendicular vectors is zero, so if we choose a normal vector \( \mathbf{m} = [a, b, c] \), we need:
$$
\mathbf{n} \cdot \mathbf{m} = 0.
$$
That is:
$$
2a + 3b + 4c = 0.
$$

We can choose any values for \( a \), \( b \), and \( c \) that satisfy this equation. For example, let’s choose \( a = 3 \), \( b = -2 \), and \( c = 0 \). Then:
$$
2(3) + 3(-2) + 4(0) = 6 - 6 + 0 = 0.
$$

$
Thus, the normal vector  \mathbf{m} = [3, -2, 0] \ is perpendicular to \ \mathbf{n} . The equation of a plane with this normal vector passing through a point \ P(1, 1, 1) \ (for example) is:
$

$$
3(x - 1) - 2(y - 1) + 0(z - 1) = 0.

Simplifying:

7.

We are given the plane equation:
$$
2x + 3y + 4z = 5,
$$
and the point \( A(1, 2, 3) \). We need to find the distance from the point \( A \) to the plane.

---

### Step 1: Formula for the distance from a point to a plane
The formula to calculate the distance \( d \) from a point \( P(x_0, y_0, z_0) \) to a plane \( ax + by + cz + d = 0 \) is given by:
$$
d = \frac{|ax_0 + by_0 + cz_0 + d|}{\sqrt{a^2 + b^2 + c^2}}.
$$
Here, the plane is given by \( 2x + 3y + 4z - 5 = 0 \), so:
- \( a = 2 \),
- \( b = 3 \),
- \( c = 4 \),
- \( d = -5 \).

The point \( A(1, 2, 3) \) has coordinates \( (x_0, y_0, z_0) = (1, 2, 3) \).

---

### Step 2: Apply the formula
Substitute the values into the formula:
$$
d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}}.
$$

First, calculate the numerator:
$$
2(1) + 3(2) + 4(3) - 5 = 2 + 6 + 12 - 5 = 15.
$$

Next, calculate the denominator:
$$
\sqrt{2^2 + 3^2 + 4^2} = \sqrt{4 + 9 + 16} = \sqrt{29}.
$$

Thus, the distance is:
$$
d = \frac{|15|}{\sqrt{29}} = \frac{15}{\sqrt{29}}.
$$

---

### Final Answer:
The distance from point \( A(1, 2, 3) \) to the plane \( 2x + 3y + 4z = 5 \) is:
$$
d = \frac{15}{\sqrt{29}}.
$$

We are given the points where the plane intersects the coordinate axes:
- \( A(2, 0, 0) \),
- \( B(0, 3, 0) \),
- \( C(0, 0, 4) \).

We need to find the equation of the plane passing through these three points.

---

### Step 1: Use the general form of the equation of a plane
The general equation of a plane is given by:
$$
Ax + By + Cz = D,
$$
where \( A \), \( B \), and \( C \) are the coefficients of the normal vector to the plane, and \( D \) is a constant.

We can substitute the coordinates of the points \( A \), \( B \), and \( C \) into the equation to find \( A \), \( B \), \( C \), and \( D \).

---

### Step 2: Find the normal vector to the plane
To find the normal vector of the plane, we first need to calculate two vectors that lie in the plane. These vectors can be found by subtracting the coordinates of the points:
- \( \mathbf{AB} = B - A = (0 - 2, 3 - 0, 0 - 0) = (-2, 3, 0) \),
- \( \mathbf{AC} = C - A = (0 - 2, 0 - 0, 4 - 0) = (-2, 0, 4) \).

Now, calculate the cross product of vectors \( \mathbf{AB} \) and \( \mathbf{AC} \) to find the normal vector to the plane:
$$
\mathbf{n} = \mathbf{AB} \times \mathbf{AC} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
-2 & 3 & 0 \\
-2 & 0 & 4
\end{vmatrix}.
$$

Expand the determinant:
$$
\mathbf{n} = \mathbf{i} \begin{vmatrix} 3 & 0 \\ 0 & 4 \end{vmatrix}
- \mathbf{j} \begin{vmatrix} -2 & 0 \\ -2 & 4 \end{vmatrix}
+ \mathbf{k} \begin{vmatrix} -2 & 3 \\ -2 & 0 \end{vmatrix}.
$$

Simplify each term:
$$
\mathbf{n} = \mathbf{i}(3 \cdot 4 - 0 \cdot 0) - \mathbf{j}((-2) \cdot 4 - (-2) \cdot 0) + \mathbf{k}((-2) \cdot 0 - (-2) \cdot 3).
$$
$$
\mathbf{n} = \mathbf{i}(12) - \mathbf{j}(-8) + \mathbf{k}(6).
$$
$$
\mathbf{n} = 12\mathbf{i} + 8\mathbf{j} + 6\mathbf{k}.
$$

Thus, the normal vector is:
$$
\mathbf{n} = (12, 8, 6).
$$

---

### Step 3: Find the equation of the plane
We now know the normal vector to the plane is \( \mathbf{n} = (12, 8, 6) \), so the equation of the plane can be written as:
$$
12x + 8y + 6z = D.
$$

To find \( D \), we substitute the coordinates of any of the points into the equation. Using point \( A(2, 0, 0) \):
$$
12(2) + 8(0) + 6(0) = D.
$$
$$
D = 24.
$$

Thus, the equation of the plane is:
$$
12x + 8y + 6z = 24.
$$

---

### Final Answer:
The equation of the plane passing through the points \( A(2, 0, 0) \), \( B(0, 3, 0) \), and \( C(0, 0, 4) \) is:
$$
12x + 8y + 6z = 24.
$$


9.

We are given two planes:
- \( P_1: x + y + z = 1 \),
- \( P_2: x = 0 \) (the \( yz \)-plane).

We need to calculate the angle between these two planes.

---

### Step 1: Find the normal vectors of the planes
The angle between two planes is determined by the angle between their normal vectors.

- The normal vector of \( P_1: x + y + z = 1 \) is \( \mathbf{n_1} = [1, 1, 1] \), because the coefficients of \( x \), \( y \), and \( z \) in the plane equation are \( 1 \), \( 1 \), and \( 1 \), respectively.
- The normal vector of \( P_2: x = 0 \) is \( \mathbf{n_2} = [1, 0, 0] \), because the plane is parallel to the \( yz \)-plane, which has a normal vector in the \( x \)-direction.

---

### Step 2: Use the formula for the angle between two vectors
The angle \( \theta \) between two vectors \( \mathbf{n_1} \) and \( \mathbf{n_2} \) is given by the formula:
$$
\cos(\theta) = \frac{\mathbf{n_1} \cdot \mathbf{n_2}}{|\mathbf{n_1}| |\mathbf{n_2}|},
$$

$
where \ \mathbf{n_1} \cdot \mathbf{n_2} \ is the dot product of the two normal vectors, and \ |\mathbf{n_1}| \ and \ |\mathbf{n_2}| \ are the magnitudes lengths of the normal vectors.
$

---

### Step 3: Calculate the dot product and magnitudes
$
- The dot product \ \mathbf{n_1} \cdot \mathbf{n_2} \ is:
$
$
\mathbf{n_1} \cdot \mathbf{n_2} = 1 \times 1 + 1 \times 0 + 1 \times 0 = 1.
$

$
- The magnitude of \ \mathbf{n_1} = [1, 1, 1] \ is:
$
$
|\mathbf{n_1}| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3}.
$

$
- The magnitude of \ \mathbf{n_2} = [1, 0, 0] \ is:
$

$
|\mathbf{n_2}| = \sqrt{1^2 + 0^2 + 0^2} = 1.
$

---

### Step 4: Calculate the angle
$
 Substitute the values into the formula for \ \cos\theta \:
$

$$
\cos(\theta) = \frac{1}{\sqrt{3} \times 1} = \frac{1}{\sqrt{3}}.
$$
$
Thus, the angle \ \theta \ is:
$
$$
\theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right).
$$

---

### Final Answer:
The angle between the plane \( x + y + z = 1 \) and the plane \( x = 0 \) (the \( yz \)-plane) is:
$$
\theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right).
$$
