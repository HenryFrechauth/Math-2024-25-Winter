# Equation of a Sphere

## Step-by-Step Explanation

We are tasked with writing the equation of a sphere with the following characteristics:
- Center: \( P(1, 2, 3) \)
- Radius: \( r = 3 \)

### General Formula for a Sphere
The general equation of a sphere with center \((x_0, y_0, z_0)\) and radius \(r\) is given by:
$$
(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 = r^2
$$

Here:
- \((x_0, y_0, z_0)\) represents the coordinates of the center.
- \(r\) is the radius of the sphere.

### Substituting the Given Values
For our problem:
- The center is \((x_0, y_0, z_0) = (1, 2, 3)\)
- The radius is \(r = 3\)

Substituting these values into the general formula:
$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 3^2
$$

### Simplifying the Equation
We know \(3^2 = 9\), so:
$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

### Final Equation of the Sphere
Thus, the equation of the sphere is:
$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

This equation represents a sphere centered at \((1, 2, 3)\) with a radius of \(3\).

# Determining Common Points of Two Spheres

## Problem
We are tasked with checking whether the two spheres given by the following equations have any common points:
$$
1. \ x^2 + y^2 + z^2 = 1 \
$$
$$
2. \ x^2 + y^2 + z^2 = 2 \
$$

## Step-by-Step Solution

### 1. General Idea
Two spheres have common points if there exist values of \(x\), \(y\), and \(z\) that satisfy both equations simultaneously.

### 2. Comparing the Two Equations
The equations of the spheres are:
$$
1. \ x^2 + y^2 + z^2 = 1 \
$$
$$
2. \ x^2 + y^2 + z^2 = 2 \
$$

From the first equation:
$$
x^2 + y^2 + z^2 = 1
$$

From the second equation:
$$
x^2 + y^2 + z^2 = 2
$$

### 3. Subtracting the Equations
Subtract the first equation from the second:
$$
(x^2 + y^2 + z^2) - (x^2 + y^2 + z^2) = 2 - 1
$$

This simplifies to:
$$
0 = 1
$$

### 4. Interpretation
The result \(0 = 1\) is a contradiction, meaning that no values of \(x\), \(y\), and \(z\) can simultaneously satisfy both equations.

### Conclusion
$
The two spheres    /x^2 + y^2 + z^2 = 1\ and \\x^2 + y^2 + z^2 = 2\ **do not have any common points**.
$

# Curve Formed by the Intersection of Two Spheres

## Problem

We are tasked with finding the curve formed by the intersection of the following two spheres:
$
1. \ x^2 + y^2 + z^2 = 1 \
$

$
2. \ (x - 1)^2 + y^2 + z^2 = 1 \
$

We also need to determine the equation of this curve.

---

## Step-by-Step Solution

### 1. Write Down the Equations of the Spheres
The first sphere is:
$$
x^2 + y^2 + z^2 = 1
$$

The second sphere is:
$$
(x - 1)^2 + y^2 + z^2 = 1
$$

### 2. Expand the Second Sphere
Expanding the second equation:
$$
(x - 1)^2 + y^2 + z^2 = 1
$$
Expands to:
$$
x^2 - 2x + 1 + y^2 + z^2 = 1
$$

Simplify:
$$
x^2 + y^2 + z^2 - 2x + 1 = 1
$$

Subtract \(1\) from both sides:
$$
x^2 + y^2 + z^2 - 2x = 0
$$

From the first sphere, we know:
$$
x^2 + y^2 + z^2 = 1
$$

### 3. Subtract the Two Equations
Subtract the expanded second equation from the first equation:
$$
(x^2 + y^2 + z^2) - (x^2 + y^2 + z^2 - 2x) = 1 - 0
$$

Simplify:
$$
2x = 1
$$

Solve for \(x\):
$$
x = \frac{1}{2}
$$

### 4. Substitute Back into the First Sphere
To find the equation of the curve, substitute \(x = \frac{1}{2}\) into the equation of the first sphere:
$$
x^2 + y^2 + z^2 = 1
$$
$
Substitute \\x = \frac{1}{2}\:
$
$$
\left(\frac{1}{2}\right)^2 + y^2 + z^2 = 1
$$

Simplify:
$$
\frac{1}{4} + y^2 + z^2 = 1
$$

$
Subtract \frac{1}{4}\ from both sides:
$

$$
y^2 + z^2 = \frac{3}{4}
$$

---

## Final Result
The curve formed by the intersection of the two spheres is a **circle** in the plane \(x = \frac{1}{2}\) with the equation:
$$
y^2 + z^2 = \frac{3}{4}
$$

This circle lies in the plane \(x = \frac{1}{2}\) and has a radius of:
$$
r = \sqrt{\frac{3}{4}} = \frac{\sqrt{3}}{2}
$$

# Tangent Plane to a Paraboloid

## Problem
We need to find the equation of the tangent plane to the paraboloid:
$$
z = (x - 1)^2 + y^2 + 1
$$
at the point \(P(1, 0, 1)\).

---

## Step-by-Step Solution

### 1. General Form of a Tangent Plane
The equation of a tangent plane to a surface \(z = f(x, y)\) at a point \((x_0, y_0, z_0)\) is given by:
$$
z - z_0 = f_x(x_0, y_0)(x - x_0) + f_y(x_0, y_0)(y - y_0)
$$
Here:
- \(f_x(x, y)\) is the partial derivative of \(f(x, y)\) with respect to \(x\).
- \(f_y(x, y)\) is the partial derivative of \(f(x, y)\) with respect to \(y\).

---

### 2. Given Surface
The given surface is:
$$
z = f(x, y) = (x - 1)^2 + y^2 + 1
$$

---

### 3. Compute Partial Derivatives
#### Partial derivative with respect to \(x\):
$$
f_x(x, y) = \frac{\partial}{\partial x}\left((x - 1)^2 + y^2 + 1\right)
$$
Differentiate:
$$
f_x(x, y) = 2(x - 1)
$$

#### Partial derivative with respect to \(y\):
$$
f_y(x, y) = \frac{\partial}{\partial y}\left((x - 1)^2 + y^2 + 1\right)
$$
Differentiate:
$$
f_y(x, y) = 2y
$$

---

### 4. Evaluate Partial Derivatives at \(P(1, 0, 1)\)
$
At \\P(1, 0, 1)\:
 \\f_x(1, 0) = 2(1 - 1) = 0\
 \\f_y(1, 0) = 2(0) = 0\
$


### 5. Equation of the Tangent Plane
The general formula for the tangent plane is:
$$
z - z_0 = f_x(x_0, y_0)(x - x_0) + f_y(x_0, y_0)(y - y_0)
$$

Substitute the values:
- \(x_0 = 1\), \(y_0 = 0\), \(z_0 = 1\)
- \(f_x(1, 0) = 0\), \(f_y(1, 0) = 0\)


Thus:
$$
z - 1 = 0(x - 1) + 0(y - 0)
$$

Simplify:
$$
z - 1 = 0
$$

Or equivalently:
$$
z = 1
$$

---

## Final Result
$
The equation of the tangent plane to the paraboloid \\z = (x - 1)^2 + y^2 + 1\ at the point \\P(1, 0, 1)\ is:
$

$$
z = 1
$$
