##Exercises




### 1. $\ y(x) = -3x + 3 \$

If we have a straight line like $\ -3x + 3 \$, the slope is the number in front of \( x \), which is \( -3 \). That’s the derivative!

**Answer:**
$$
\frac{dy}{dx} = -3
$$

---

### 2. $\ y(x) = \pi x + \sin(1) \$

For $\ \pi x \, \ \pi \$ is just a number, so the derivative is

$\ \pi \$.

The other part, $\ \sin(1) \$, is a constant (it doesn’t change), so its derivative is \( 0 \).

**Answer:**
$$
\frac{dy}{dx} = (\pi x + \sin(1))'=(\pi * x)' + sin(1)'= \pi*(x)' +0= \pi *1=\pi
$$

---

### 3. $\ y(x) = 4 + \sin(2) \$

Here, both \( 4 \) and $\ \sin(2) \$ are constants—they don’t have an \( x \) in them. If something doesn’t change, its derivative is \( 0 \).

**Answer:**
$$
\frac{dy}{dx} = \ y(x)' =( 4 + \sin(2))'=4'+ sin(2)'= 0 + 0 = 0
$$

---

### 4. $\ y(x) = 2x^3 - 3x^2 + 8x - 9 \$

To find the derivative, look at each part:
- For \( 2x^3 \), bring down the 3 and multiply: $\ 3 \cdot 2 = 6 \$. Then
 lower the power to \( 2 \): \( 6x^2 \).

- For \( -3x^2 \), bring down the 2: \( 2 \cdot -3 = -6 \). Lower the power to \( 1 \): \( -6x \).

- For \( 8x \), the derivative is just \( 8 \) (it’s like the slope).

- For \( -9 \), it’s a constant, so the derivative is \( 0 \).

**Answer:**
$$
\frac{dy}{dx} = 6x^2 - 6x + 8
$$

---

### 5. $\ y(x) = 6x^{1/3} \$

This looks tricky, but it’s just like the power rule! Bring the
$\ \frac{1}{3} \$ down and multiply by

$\( 6 \): \ \frac{1}{3} \cdot 6 = 2 \$.
Then subtract 1 from the power:

$\ \frac{1}{3} - 1 = -\frac{2}{3} \$.

**Answer:**
$$
\frac{dy}{dx} = \frac{2}{\sqrt[3]{x^2}}
$$

---

### 6. $\ y(x) = \sqrt{x} \$

The square root of \( x \) is the same as $\ x^{1/2} \$. Use the power rule:

- Bring the $\ \frac{1}{2} \ $
down and subtract 1 from the power:

$\ \frac{1}{2} - 1 = -\frac{1}{2} \$.

**Answer:**
$$
\frac{dy}{dx} = \frac{1}{2\sqrt{x}}
$$

---

### 7. $\ y(x) = \cos(x) + \sin(x) \$

- The derivative of \( cos(x) \) is \( -sin(x) \).
- The derivative of \( sin(x) \) is \( cos(x) \).

**Answer:**
$$
\frac{dy}{dx} = -\sin(x) + \cos(x)
$$

---

### 8. $\ y(x) = 2\sin(x)\cos(x) \$

This is a product of two things: \( \sin(x) \) and \( \cos(x) \). Use the product rule:

1. Differentiate \( sin(x) \): it becomes \( cos(x) \).

2. Differentiate \( cos(x) \): it becomes \( -sin(x) \).

3. Multiply and add together:
   - First part: $\ 2 \cdot \cos(x) \cdot \cos(x) \$

   - Second part: $\ 2 \cdot \sin(x) \cdot (-\sin(x)) \$

Simplify:
$$
\cos^2(x) - \sin^2(x)
$$

**Answer:**
$$
\frac{dy}{dx} = 2\cos(2x)
$$

---

### 9. $\ y(x) = x\sin(x) \$

Here, we multiply \( x \) by \( \sin(x) \). Use the product rule:

1. Differentiate \( x \): it becomes \( 1 \).

2. Differentiate $\ \sin(x) \: it becomes \ \cos(x) \$.

3. Combine: $\ (1)(\sin(x)) + (x)(\cos(x)) \$.

**Answer:**
$$
\frac{dy}{dx} = \sin(x) + x\cos(x)
$$

---

### 10. $\ y(x) = (x+1)(x+1) \$

This is just $\ (x+1)^2 \$. Use the chain rule:

1. Differentiate the outer part $\ (u^2) \: it becomes \ 2u \$.

2. Inside \( u = x+1 \): the derivative is \( 1 \).

3. Combine: $\ 2(x+1) \cdot 1 \$.

**Answer:**
$$
\frac{dy}{dx} = 2(x+1)
$$

---

### 11. $\ y(x) = \frac{x}{x+1} \$

This is a fraction, so we use the quotient rule:
1. Differentiate the top (\( x \)): it becomes \( 1 \).
2. Differentiate the bottom (\( x+1 \)): it becomes \( 1 \).
3. Apply the rule:
   - Multiply the top by the derivative of the bottom: $\ 1 \cdot (x+1) \$.

   - Subtract the bottom times the derivative of the top: $\ x \cdot 1 \$.
   - Divide by the square of the bottom.

Simplify:
$$
\frac{1}{(x+1)^2}
$$

**Answer:**
$$
\frac{dy}{dx} = \frac{1}{(x+1)^2}
$$

---

### 12. $\ y(x) = (x+1)e^x \$

This is a product, so we use the product rule:
1. Differentiate \( x+1 \): it becomes \( 1 \).
2. Differentiate \( e^x \): it stays  $\ e^x \$.
3. Combine:
   - First part: $\ (1)(e^x) \$.
   - Second part: $\ (x+1)(e^x) \$.

Simplify:
$$
e^x(x+2)
$$

**Answer:**
$$
\frac{dy}{dx} = e^x(x+2)
$$

---

### 13. $\ y(x) = \sin(x^2) \$

This is a chain rule problem:
1. The outer part is $\ \sin(u) \, which becomes \ \cos(u) \$.
2. The inner part is \( x^2 \), which becomes \( 2x \).
3. Combine: \( \cos(x^2) \cdot 2x \).

**Answer:**
$$
\frac{dy}{dx} = 2x\cos(x^2)
$$

---

### 14.  y(x) = $e^{-2x} \$

This is another chain rule:
1. The outer part is \( e^u \), which stays \( e^u \).
2. The inner part is \( -2x \), which becomes \( -2 \).
3. Combine: $\ e^{-2x} \cdot (-2) \$.

**Answer:**
$$
\frac{dy}{dx} = -2e^{-2x}
$$

# Explanation of $\ \frac{d}{dx} \ln(\sin(x)) = \cot(x) \$

---

### Step 1: Start with what we are solving

We want to figure out how $\ \ln(\sin(x)) \$ changes when \( x \) changes. In math, we call this "taking the derivative."

So, our goal is to find:

$$
\frac{d}{dx} \ln(\sin(x)).
$$

---

### Step 2: The rule for the natural logarithm

When you take the derivative of a logarithm (written as $\ \ln \$), there’s a special rule:

$$
\frac{d}{dx} \ln(\text{something}) = \frac{1}{\text{that something}} \cdot \text{the derivative of that something}.
$$

In our case, the "something" is $\ \sin(x) \$. So, we can write:

$$
\frac{d}{dx} \ln(\sin(x)) = \frac{1}{\sin(x)} \cdot \frac{d}{dx}(\sin(x)).
$$

---

### Step 3: What’s the derivative of $\ \sin(x) \$?

Now, we need to figure out the derivative of \( sin(x) \). Remember, the derivative of \( sin(x) \) is \( cos(x) \). So we replace it in the formula:

$$
\frac{d}{dx} \ln(\sin(x)) = \frac{1}{\sin(x)} \cdot \cos(x).
$$

---

### Step 4: Simplify the expression

When we multiply $\ \frac{1}{\sin(x)} \ by \ \cos(x) \$, it becomes:

$$
\frac{\cos(x)}{\sin(x)}.
$$

This is a special function in math called $\ \cot(x) \ (cotangent)$. So we can simplify:

$$
\frac{d}{dx} \ln(\sin(x)) = \cot(x).
$$

---

### Step 5: Final Answer

The derivative of $\ \ln(\sin(x)) \$ is:

$$
\frac{d}{dx} \ln(\sin(x)) = \cot(x).
$$

# Verifying $\ f''(x) = -f(x) \ for \ f(x) = \cos(x) \$



---

## Step 1: What does $\ f(x) = \cos(x) \$ mean?

This is just a fancy way of saying that we have a function called \( f(x) \), and it equals $\ \cos(x) \$.

The function $\ \cos(x) \$ takes an angle \( x \) and gives us a number. It’s like a wave that goes up and down, repeating forever, just like the shape of ocean waves.

---

## Step 2: What is a derivative?
A derivative is like asking:
- **"How fast is this wave going up or down at a certain point?"**

For example:
- If the wave is going up quickly, the derivative is a big positive number.
- If the wave is flat, the derivative is \( 0 \).
- If the wave is going down quickly, the derivative is a big negative number.

Now, we’ll find the derivatives of $\ f(x) = \cos(x) \$ step by step.

---

## Step 3: First derivative
The first derivative of $\ f(x) = \cos(x) \$ is:
$$
f'(x) = -\sin(x)
$$

This tells us how fast $\ \cos(x) \$ is going up or down at each point.

---

## Step 4: Second derivative
Now, let’s find the derivative of $\ f'(x) = -\sin(x) \$

- The derivative of $\ \sin(x) \ is \ \cos(x) \$.
- Since there’s a negative sign in front, we get:
$$
f''(x) = -\cos(x)
$$

So the second derivative is:
$$
f''(x) = -\cos(x)
$$

---

## Step 5: Compare with \( -f(x) \)
Remember, we started with:
$$
f(x) = \cos(x)
$$

If we multiply \( f(x) \) by \( -1 \), we get:
$$
-f(x) = -\cos(x)
$$

But wait! That’s exactly the same as \( f''(x) \):
$$
f''(x) = -f(x)
$$

---

## Step 6: Final answer
We’ve shown that the second derivative of $\ f(x) = \cos(x) \$ is the same as \( -f(x) \).

# Using L'Hôpital's Rule to Find Improper Limits

---

## 1. $\ \lim_{x \to 0} \frac{\sin(x)}{x} \$

### Step 1: Check if L'Hôpital's Rule applies
As $\ x \to 0 \$:

- $\ \sin(x) \to 0 \$

- $\ x \to 0 \$

The limit is of the indeterminate form \( \frac{0}{0} \), so we can apply L'Hôpital's Rule.

### Step 2: Differentiate numerator and denominator
- Derivative of $\ \sin(x) \ is \ \cos(x) \$.
- Derivative of \( x \) is \( 1 \).

Apply L'Hôpital's Rule:
$$
\lim_{x \to 0} \frac{\sin(x)}{x} = \lim_{x \to 0} \frac{\cos(x)}{1}
$$

Substitute \( x = 0 \):
$$
\cos(0) = 1
$$

**Answer**:
$$
\lim_{x \to 0} \frac{\sin(x)}{x} = 1
$$

---

## 2. $\ \lim_{x \to \infty} \frac{\ln(x)}{x} \$

### Step 1: Check if L'Hôpital's Rule applies
As $\ x \to \infty \$:

- $\ \ln(x) \to \infty \$

- $\ x \to \infty \$

The limit is of the indeterminate form $\ \frac{\infty}{\infty} \$, so we can apply L'Hôpital's Rule.

### Step 2: Differentiate numerator and denominator
- Derivative of $\ \ln(x) \ is \ \frac{1}{x} \$.

- Derivative of \( x \) is \( 1 \).

Apply L'Hôpital's Rule:
$$
\lim_{x \to \infty} \frac{\ln(x)}{x} = \lim_{x \to \infty} \frac{\frac{1}{x}}{1}
$$

Simplify:
$$
\lim_{x \to \infty} \frac{1}{x} = 0
$$

**Answer**:
$$
\lim_{x \to \infty} \frac{\ln(x)}{x} = 0
$$

---

## 3. $\ \lim_{x \to \infty} \frac{\exp(x)}{x} \$

### Step 1: Check if L'Hôpital's Rule applies
As $\ x \to \infty \$:

- $\ \exp(x) \to \infty \$

- $\ x \to \infty \$

The limit is of the indeterminate form $\ \frac{\infty}{\infty} \$, so we can apply L'Hôpital's Rule.

### Step 2: Differentiate numerator and denominator
- Derivative of $\ \exp(x) \ is \ \exp(x) \$.

- Derivative of \( x \) is \( 1 \).

Apply L'Hôpital's Rule:
$$
\lim_{x \to \infty} \frac{\exp(x)}{x} = \lim_{x \to \infty} \frac{\exp(x)}{1}
$$

Simplify:
$$
\lim_{x \to \infty} \exp(x) = \infty
$$

**Answer**:
$$
\lim_{x \to \infty} \frac{\exp(x)}{x} = \infty
$$

## Understanding Velocity and Acceleration of a Particle

---

### The Problem:
We’re tracking a particle as it moves. Its position at any time \( t \) is given by:  
$$
x(t) = 3t^2 - 6t + 1
$$

We need to figure out:
1. How **fast** the particle is moving (**velocity**: \( V(t) \)).
2. How the speed is **changing** over time (**acceleration**: \( a(t) \)).
3. Find both velocity and acceleration at \( t = 2 \).

---

### Step 1: What is Velocity?
Velocity is just a fancy way of saying, **"How fast is the particle moving?"**

We find velocity by taking the **derivative** of the position equation \( x(t) \).  
Think of this as checking how much \( x(t) \) changes as time passes.

Starting with:
$$
x(t) = 3t^2 - 6t + 1
$$

1. The derivative of \( 3t^2 \) is \( 6t \).  
2. The derivative of \( -6t \) is \( -6 \).  
3. The derivative of \( 1 \) is \( 0 \).

So, the velocity is:
$$
V(t) = 6t - 6
$$

---

### Step 2: What is Acceleration?
Acceleration is just a fancy way of saying, **"How fast is the velocity changing?"**

We find acceleration by taking the **derivative** of the velocity equation \( V(t) \).  

From:
$$
V(t) = 6t - 6
$$

1. The derivative of \( 6t \) is \( 6 \).  
2. The derivative of \( -6 \) is \( 0 \).  

So, the acceleration is:
$$
a(t) = 6
$$

This means the acceleration is always \( 6 \), no matter the time.

---

### Step 3: What Happens at \( t = 2 \)?
Now, let’s find the velocity and acceleration when \( t = 2 \).

1. **Velocity at \( t = 2 \):**
$$
V(2) = 6(2) - 6 = 12 - 6 = 6
$$

2. **Acceleration at \( t = 2 \):**
$$
a(2) = 6
$$

---

### Final Answer:
1. At \( t = 2 \), the **velocity** is:
$$
V(2) = 6 \, \text{units per second.}
$$

2. The **acceleration** is:
$$
a(2) = 6 \, \text{units per second squared.}
$$
