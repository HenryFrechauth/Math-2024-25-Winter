# How Many Units Maximize Profit?

We are trying to figure out how many units (\(u\)) need to be sold to make the most profit using this profit formula:

$$
P(u) = -2u^2 + 50u - 300
$$

---

## Step 1: Understand the Equation
- The formula tells us how profit (\(P\)) depends on the number of units sold (\(u\)).
- The shape of this formula is a **parabola**.
- Since the number in front of \(u^2\) (called \(a\)) is negative (\(-2\)), the parabola is shaped like a hill. This means it has a **highest point**.
- This highest point is where the profit is the greatest.

---

## Step 2: Formula for the Highest Point
The highest point of a parabola can be found using this formula:

$$
u = -\frac{b}{2a}
$$

In our equation:
- a = -2 (the number in front of $\ (u^2\ ) )$,

- b = 50 (the number in front of \(u\)).

---

## Step 3: Plug in the Numbers
We substitute \(a = -2\) and \(b = 50\) into the formula:

$$
u = -\frac{b}{2a}
$$

Substitute the values:

$$
u = -\frac{50}{2(-2)}
$$

---

## Step 4: Simplify Step by Step
1. First, calculate \(2 \times -2\):

$$
2 \times -2 = -4
$$

   So now the equation becomes:

$$
u = -\frac{50}{-4}
$$

2. Next, simplify the fraction:
   - Divide \(-50\) by \(-4\). The two negative signs cancel out, making the result positive:

$$
u = \frac{50}{4} = 12.5
$$

---

## Step 5: Write the Result
The number of units (\(u\)) that gives the maximum profit is:

$$
u = 12.5
$$

---

## Step 6: Check the Maximum Profit
Now that we know \(u = 12.5\), let’s calculate the maximum profit by substituting \(u = 12.5\) into the original equation:

$$
P(u) = -2u^2 + 50u - 300
$$

Substitute \(u = 12.5\):

$$
P(12.5) = -2(12.5)^2 + 50(12.5) - 300
$$

---

### Step 7: Simplify Step by Step
1. Calculate \(12.5^2\):
   $$
   12.5^2 = 156.25
   $$

2. Multiply by \(-2\):
   $$
   -2 \times 156.25 = -312.5
   $$

3. Multiply $\ ( 50 \times 12.5\ )$ :
   $$
   50 \times 12.5 = 625
   $$

4. Subtract 300:
   $$
   -300 = -300
   $$

Now combine all the results:
$$
P(12.5) = -312.5 + 625 - 300
$$

Simplify:
$$
P(12.5) = 12.5
$$

---

## Step 8: Final Answer
The maximum profit is:

$$
P = 12.5
$$

The number of units sold for the maximum profit is:

$$
u = 12.5
$$

# How to Find the Largest Rectangle with 10 Meters of String

We have **10 meters of string**, and we want to make the biggest rectangle possible. Let’s figure it out step by step!

---

## Step 1: Perimeter of a Rectangle
The **perimeter** is the total length of the string needed to go around the rectangle. The formula is:

$$
P = 2L + 2W
$$

Here:
- \(L\) is the length of the rectangle,
- \(W\) is the width of the rectangle.

We know the perimeter is **10 meters**, so:

$$
2L + 2W = 10
$$

---

## Step 2: Simplify the Equation
Divide everything by 2 to make it easier:

$$
L + W = 5
$$

This means that if you add the length and width together, you get **5 meters**.

---

## Step 3: Write the Area Formula
The **area** of a rectangle tells us how much space is inside. The formula is:

$$
A = L \times W
$$

From Step 2, we know that \(W = 5 - L\). Substitute this into the area formula:

$$
A = L \times (5 - L)
$$

---

## Step 4: Simplify the Area Formula
Distribute \(L\) to simplify:

$$
A = 5L - L^2
$$

This is the formula for the area.

---

## Step 5: Find the Maximum Area
The formula \(A = 5L - L^2\) is a parabola. Since it opens downward (because of the $\ (-L^2\ )$ ), the highest point will give us the largest area.

The formula to find the highest point (vertex) is:

$$
L = -\frac{b}{2a}
$$

Here:
- a = -1 (the coefficient of $\ (-L^2\ )$),

- b = 5 (the coefficient of \(5L\)).

Substitute the values:

$$
L = -\frac{5}{2(-1)} = \frac{5}{2} = 2.5
$$

So, the **Length** is:

$$
L = 2.5 \, \text{meters}
$$

---

## Step 6: Find the Width
We know:

$$
W = 5 - L
$$

Substitute \(L = 2.5\):

$$
W = 5 - 2.5 = 2.5
$$

The **Width** is:

$$
W = 2.5 \, \text{meters}
$$

---

## Step 7: The Rectangle is a Square
The rectangle with the largest area is actually a **square** with sides of:

$$
2.5 \, \text{meters}.
$$

---

## Step 8: Calculate the Maximum Area
The **area** of the rectangle is:

$$
A = L \times W
$$

Substitute \(L = 2.5\) and \(W = 2.5\):

$$
A = 2.5 \times 2.5 = 6.25
$$

---

## Final Answer:
The rectangle with the largest area is a square with sides of:

$$
2.5 \, \text{meters each}.
$$

The **maximum area** is:

$$
6.25 \, \text{square meters}.
$$

# Finding the Extremum of $\ ( f(x) = x^2 + 3x - 5 \ )$

We are going to find the lowest point (minimum) of the function $\ ( f(x) = x^2 + 3x - 5 \ )$ step by step.

---

## Step 1: Take the derivative of \( f(x) \)

The derivative tells us the slope of the function. Using the rule:
- The derivative of $\ ( x^2 \ )$ is \( 2x \),
- The derivative of \( 3x \) is \( 3 \),
- The derivative of a constant like \( -5 \) is \( 0 \),

we get:

$$
f'(x) = 2x + 3
$$

---

## Step 2: Find where the slope is zero

To find the flat spot, set \( f'(x) = 0 \):

$$
2x + 3 = 0
$$

Solve for \( x \):
1. Subtract \( 3 \) from both sides:
   $$
   2x = -3
   $$
2. Divide by \( 2 \):
   $$
   x = -\frac{3}{2}
   $$

So the flat spot happens at $\ ( x = -\frac{3}{2} \ )$.

---

## Step 3: Determine if it’s a minimum or maximum

To check if this point is the bottom (minimum) or the top (maximum) of the curve, we use the second derivative. The second derivative of \( f(x) \) is:

$$
f''(x) = \frac{d}{dx}(2x + 3) = 2
$$

- If \( f''(x) > 0 \), the curve is a smile ($\ ( \smile \ )$) and has a minimum.
- If \( f''(x) < 0 \), the curve is a frown ($\ ( \frown \ )$) and has a maximum.

Since \( f''(x) = 2 > 0 \), the curve is a smile, so $\ ( x = -\frac{3}{2} \ )$ is a **minimum**.

---

## Step 4: Find the value of \( f(x) \) at $\ ( x = -\frac{3}{2} \ )$

To find how low the curve goes, substitute $\ ( x = -\frac{3}{2} \ )$ into the original function $\ ( f(x) = x^2 + 3x - 5 \ )$:

$$
f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5
$$

Step by step:
1. Square $\ ( -\frac{3}{2} \ )$:

   $$
   \left(-\frac{3}{2}\right)^2 = \frac{9}{4}
   $$

2. Multiply $\ ( 3 \times -\frac{3}{2} \ )$:
   $$
   3 \times -\frac{3}{2} = -\frac{9}{2}
   $$
3. Combine everything:
   $$
   f\left(-\frac{3}{2}\right) = \frac{9}{4} - \frac{18}{4} - \frac{20}{4}
   $$
   $$
   f\left(-\frac{3}{2}\right) = \frac{9 - 18 - 20}{4} = \frac{-29}{4}
   $$

---

## Final Answer

The minimum value of the function is:

$$
f(x) = \frac{-29}{4} \text{ at } x = -\frac{3}{2}
$$

So the minimum point is:

$$
\left( -\frac{3}{2}, \frac{-29}{4} \right)
$$

# Finding the Extremum of $\ ( f(x) = \frac{(x + 1)^2}{x - 1} \ )$

We’re going to find the highest and lowest points (extrema) of the function step by step.

---

## Step 1: Simplify the function
The numerator $\ ( x^2 + 2x + 1 \ )$ can be factored as $\ ( (x + 1)^2 \ )$, so the function becomes:

$$
f(x) = \frac{(x + 1)^2}{x - 1}
$$

---

## Step 2: Find the derivative
We use the **quotient rule** to differentiate \( f(x) \). The quotient rule says:

$$
f'(x) = \frac{u'(x)v(x) - u(x)v'(x)}{[v(x)]^2}
$$

Here:
- $\ ( u(x) = (x + 1)^2 \ ), so \ ( u'(x) = 2(x + 1) \ )$,
- $\ ( v(x) = x - 1 \ ), so \ ( v'(x) = 1 \ )$.

Substitute into the quotient rule:

$$
f'(x) = \frac{[2(x + 1)](x - 1) - (x + 1)^2(1)}{(x - 1)^2}
$$

---

## Step 3: Simplify the derivative
Let’s simplify the numerator step by step:

1. Expand $\ ( [2(x + 1)](x - 1) \ )$:
   $$
   [2(x + 1)](x - 1) = 2(x^2 - 1) = 2x^2 - 2
   $$

2. Expand $\ ( (x + 1)^2 \ )$:
   $$
   (x + 1)^2 = x^2 + 2x + 1
   $$

3. Subtract the second part from the first:
   $$
   \text{Numerator: } (2x^2 - 2) - (x^2 + 2x + 1)
   $$

   Simplify:
   $$
   \text{Numerator: } x^2 - 2x - 3
   $$

So the derivative is:

$$
f'(x) = \frac{x^2 - 2x - 3}{(x - 1)^2}
$$

---

## Step 4: Find critical points
Critical points occur when \( f'(x) = 0 \) or \( f'(x) \) is undefined.

### Step 4.1: Solve \( f'(x) = 0 \)
The numerator of \( f'(x) \) is \( x^2 - 2x - 3 \). Set it equal to 0:

$$
x^2 - 2x - 3 = 0
$$

Factorize the quadratic:

$$
x^2 - 2x - 3 = (x - 3)(x + 1)
$$

So:

$$
x = 3 \quad \text{or} \quad x = -1
$$

### Step 4.2: Check where \( f'(x) \) is undefined
The denominator $\ ( (x - 1)^2 \ )$ is undefined when \( x = 1 \). So, \( x = 1 \) is a **vertical asymptote**.

---

## Step 5: Determine the nature of the critical points
To see if \( x = 3 \) and \( x = -1 \) are maxima or minima, examine the behavior of \( f'(x) \) around these points. This tells us if the function is increasing or decreasing.

---

## Step 6: Find the value of \( f(x) \) at the critical points
Let’s substitute \( x = 3 \) and \( x = -1 \) into the original function to find their heights.

### For \( x = 3 \):
Substitute \( x = 3 \) into $\ ( f(x) = \frac{(x + 1)^2}{x - 1} \ )$:

$$
f(3) = \frac{(3 + 1)^2}{3 - 1} = \frac{4^2}{2} = \frac{16}{2} = 8
$$

### For \( x = -1 \):
Substitute \( x = -1 \) into $\ ( f(x) = \frac{(x + 1)^2}{x - 1} \ )$:

$$
f(-1) = \frac{(-1 + 1)^2}{-1 - 1} = \frac{0^2}{-2} = 0
$$

---

## Final Answer

- The function has a **minimum** at \( x = -1 \), with \( f(-1) = 0 \).
- The function has a **maximum** at \( x = 3 \), with \( f(3) = 8 \).
- The function has a **vertical asymptote** at \( x = 1 \), where it isn’t defined.
