### Step 1: What does $\ ( y'(x) = y(x) \ )$ mean?
The equation $\ ( y'(x) = y(x) \ )$ tells us that the function $\ ( y(x) \ )$ grows in a special way: the bigger $\ ( y(x) \ )$ gets, the faster it grows. This is like a snowball rolling down a hill—the bigger it gets, the faster it rolls!

---

### Step 2: Separate the variables
To solve this, we need to separate the \( y \) and \( x \) parts of the equation. We can rewrite the equation as:

$$
\frac{dy}{dx} = y
$$

This means the change in $\ ( y \ ) (\ ( dy \ ))$ divided by the change in \( x \) (\( dx \)) is equal to \( y \). To separate \( y \) and \( x \), we move all the \( y \) terms to one side and the \( x \) terms to the other:

$$
\frac{dy}{y} = dx
$$

---

### Step 3: Integrate both sides
Now we integrate both sides of the equation. Integration is like adding up all the tiny changes to find the total.

- On the left side, we integrate $\ ( \frac{dy}{y} \ )$. This is a special integral, and it gives us $\ ( \ln|y| \ )$ (the natural logarithm of \( y \)).
- On the right side, we integrate \( dx \), which is just \( x \).

So, we get:

$$
\ln|y| = x + C
$$

Here, \( C \) is a constant that we add because we don’t know the exact starting point of \( y \).

---

### Step 4: Solve for \( y \)
Now we need to solve for \( y \). To do this, we use the fact that $\ ( \ln|y| \ )$ is the natural logarithm of \( y \). To "undo" the natural logarithm, we use the exponential function \( e \). So, we exponentiate both sides:

$$
e^{\ln|y|} = e^{x + C}
$$

Simplifying this:

$$
|y| = e^x \cdot e^C
$$

Since $\ ( e^C \ )$ is just another constant, we can write:

$$
|y| = C_1 e^x
$$

where $\ ( C_1 = e^C \ )$.

---

### Step 5: Remove the absolute value
The absolute value \( |y| \) means \( y \) could be positive or negative. So, we write:

$$
y = \pm C_1 e^x
$$

But $\ ( \pm C_1 \ )$ is just another constant, so we can rewrite this as:

$$
y(x) = C_1 e^x
$$

where $\ ( C_1 \ )$ is a constant that can be positive or negative.

---

### Step 6: Final answer
The solution to the equation \( y'(x) = y(x) \) is:

$$
y(x) = C_1 e^x
$$

This means \( y(x) \) grows exponentially as \( x \) increases. The constant $\ ( C_1 \ )$ depends on the starting value of \( y(x) \). For example, if $\ ( y(0) = 1 \ )$, then $\ ( C_1 = 1 \ )$, and the solution becomes:

$$
y(x) = e^x
$$

---

The function $\ ( y(x) = C_1 e^x \ )$ is the solution to the equation $\ ( y'(x) = y(x) \ )$. It grows exponentially, just like a snowball rolling down a hill!

##2.1

### Step 1: Understand the equation
The equation $\ ( \frac{dy}{dx} = \frac{x}{y} \ )$ tells us how \( y \) changes as \( x \) changes. Specifically, the rate of change of $\ ( y \ ) (\ ( \frac{dy}{dx} \ ))$ is equal to $\ ( \frac{x}{y} \ )$. This means:
- If \( x \) is positive and \( y \) is positive, \( y \) increases as \( x \) increases.
- If \( x \) is negative and \( y \) is positive, \( y \) decreases as \( x \) increases.

Our goal is to find \( y(x) \), the function that satisfies this relationship.

---

### Step 2: Separate the variables
To solve this equation, we use the **method of separation of variables**. This means we want to rearrange the equation so that all terms involving \( y \) are on one side and all terms involving \( x \) are on the other side.

Start with the original equation:

$$
\frac{dy}{dx} = \frac{x}{y}
$$

Multiply both sides by \( y \) to move \( y \) to the left side:

$$
y \cdot \frac{dy}{dx} = x
$$

Now, multiply both sides by \( dx \) to move \( dx \) to the right side:

$$
y \, dy = x \, dx
$$

Now the equation is **separated**:
- The left side has only \( y \) and \( dy \).
- The right side has only \( x \) and \( dx \).

---

### Step 3: Integrate both sides
Now that the variables are separated, we integrate both sides. Integration is the process of finding the total change from small changes (like adding up tiny pieces).

Integrate the left side with respect to \( y \):

$$
\int y \, dy
$$

Integrate the right side with respect to \( x \):

$$
\int x \, dx
$$

The integrals are:

$$
\int y \, dy = \frac{1}{2} y^2 + C_1
$$

$$
\int x \, dx = \frac{1}{2} x^2 + C_2
$$

Here, $\ ( C_1 \ ) and \ ( C_2 \ )$ are constants of integration. When we combine them, we can write:

$$
\frac{1}{2} y^2 = \frac{1}{2} x^2 + C
$$

where $\ ( C = C_2 - C_1 \ )$ is a single constant.

---

### Step 4: Solve for \( y \)
Now we solve for \( y \). Start by multiplying both sides by 2 to eliminate the fraction:

$$
y^2 = x^2 + 2C
$$

Let $\ ( 2C = C_1 \ )$, where $\ ( C_1 \ )$ is a new constant. Then:

$$
y^2 = x^2 + C_1
$$

To solve for \( y \), take the square root of both sides:

$$
y = \pm \sqrt{x^2 + C_1}
$$

The $\ ( \pm \ )$ means \( y \) can be either positive or negative, depending on the problem’s context.

---

### Step 5: General solution
The general solution to the differential equation is:

$$
y(x) = \pm \sqrt{x^2 + C_1}
$$

where $\ ( C_1 \ )$ is an arbitrary constant. This constant is determined by the **initial condition** of the problem (e.g., the value of \( y \) at \( x = 0 \)).

---

### Step 6: Example (if initial condition is given)
Suppose we are given the initial condition \( y(0) = 2 \). This means when \( x = 0 \), \( y = 2 \). We can use this to find the value of $\ ( C_1 \ )$.

Substitute \( x = 0 \) and \( y = 2 \) into the general solution:

$$
2 = \pm \sqrt{0^2 + C_1}
$$

Simplify:

$$
2 = \pm \sqrt{C_1}
$$

Since $\ ( \sqrt{C_1} \ )$ is always non-negative, we take the positive root:

$$
\sqrt{C_1} = 2 \implies C_1 = 4
$$

Thus, the particular solution is:

$$
y(x) = \sqrt{x^2 + 4}
$$

---

### Step 7: Verify the solution
To ensure our solution is correct, we can substitute $\ ( y(x) = \sqrt{x^2 + 4} \ )$ back into the original differential equation.

First, compute $\ ( \frac{dy}{dx} \ )$:

$$
y(x) = \sqrt{x^2 + 4} = (x^2 + 4)^{1/2}
$$

Using the chain rule, the derivative is:

$$
\frac{dy}{dx} = \frac{1}{2} (x^2 + 4)^{-1/2} \cdot 2x = \frac{x}{\sqrt{x^2 + 4}}
$$

Now,  $\ ( \frac{x}{y} \ )$:

$$
\frac{x}{y} = \frac{x}{\sqrt{x^2 + 4}}
$$

Since $\ ( \frac{dy}{dx} = \frac{x}{y} \ )$, the solution satisfies the original equation.

---

###  Answer:
The general solution to the differential equation $\ ( \frac{dy}{dx} = \frac{x}{y} \ )$ is:

$$
y(x) = \pm \sqrt{x^2 + C_1}
$$

where $\ ( C_1 \ )$ is a constant determined by initial conditions. If \( y(0) = 2 \), the particular solution is:

$$
y(x) = \sqrt{x^2 + 4}
$$

##3.1

### Step 1: Rewrite the equation
The given differential equation is:

$$
y''(x) + y'(x) = 0
$$

This is a **linear second-order homogeneous differential equation** with constant coefficients. To solve it, we first find the characteristic equation.

---

### Step 2: Find the characteristic equation
Assume a solution of the form $\ ( y(x) = e^{rx} \ )$, where $\ ( r \ )$ is a constant. Substitute $\ ( y(x) = e^{rx} \ )$ into the differential equation:

$$
y''(x) + y'(x) = 0
$$

First,  $\ ( y'(x) \ ) and \ ( y''(x) \ )$:

$$
y'(x) = r e^{rx}, \quad y''(x) = r^2 e^{rx}.
$$

Substitute these into the equation:

$$
r^2 e^{rx} + r e^{rx} = 0.
$$

Factor out $\ ( e^{rx} \ )$:

$$
e^{rx} (r^2 + r) = 0.
$$

Since $\ ( e^{rx} \neq 0 \ )$, the characteristic equation is:

$$
r^2 + r = 0.
$$

---

### Step 3: Solve the characteristic equation
Solve the characteristic equation $\ ( r^2 + r = 0 \ )$:

$$
r(r + 1) = 0.
$$

This gives two roots:

$$
r = 0 \quad \text{and} \quad r = -1.
$$

---

### Step 4: Write the general solution
The general solution to the differential equation is a linear combination of the solutions corresponding to the roots of the characteristic equation. For \( r = 0 \) and \( r = -1 \), the general solution is:

$$
y(x) = C_1 e^{0x} + C_2 e^{-x},
$$

where \( C_1 \) and \( C_2 \) are constants. Simplify:

$$
y(x) = C_1 + C_2 e^{-x}.
$$

---

### Step 5: Apply the boundary conditions
We now use the boundary conditions to find the constants $\ ( C_1 \ ) and \ ( C_2 \ )$.

#### Boundary condition 1: \( y(0) = 2 \)
Substitute \( x = 0 \) and \( y(0) = 2 \) into the general solution:

$$
2 = C_1 + C_2 e^{0}.
$$

Simplify:

$$
2 = C_1 + C_2.
$$

#### Boundary condition 2: \( y'(0) = -1 \)
First, compute \( y'(x) \) from the general solution:

$$
y(x) = C_1 + C_2 e^{-x}.
$$

Differentiate with respect to \( x \):

$$
y'(x) = -C_2 e^{-x}.
$$

Substitute \( x = 0 \) and \( y'(0) = -1 \):

$$
-1 = -C_2 e^{0}.
$$

Simplify:

$$
-1 = -C_2 \implies C_2 = 1.
$$

---

### Step 6: Solve for \( C_1 \)
From the first boundary condition, we have:

$$
2 = C_1 + C_2.
$$

Substitute $\ ( C_2 = 1 \ )$:

$$
2 = C_1 + 1 \implies C_1 = 1.
$$

---

### Step 7: Write the particular solution
Substitute $\ ( C_1 = 1 \ ) and \ ( C_2 = 1 \ )$ into the general solution:

$$
y(x) = 1 + e^{-x}.
$$

---

### Final Answer:
The solution to the differential equation $\ ( y''(x) + y'(x) = 0 \ )$ with the boundary conditions \( y(0) = 2 \) and \( y'(0) = -1 \) is:

$$
y(x) = 1 + e^{-x}.
$$

---

### Verification:
To verify the solution, compute \( y'(x) \) and \( y''(x) \):

$$
y(x) = 1 + e^{-x},
$$

$$
y'(x) = -e^{-x},
$$

$$
y''(x) = e^{-x}.
$$

Substitute into the differential equation:

$$
y''(x) + y'(x) = e^{-x} + (-e^{-x}) = 0.
$$

The solution satisfies the differential equation. Additionally:
- $\ ( y(0) = 1 + e^{0} = 2 \ )$,
- $\ ( y'(0) = -e^{0} = -1 \ )$.



##3.2

### Step 1: Rewrite the equation
The given differential equation is:

$$
y''(x) - y(x) = 0
$$

This is a **linear second-order homogeneous differential equation** with constant coefficients. To solve it, we first find the characteristic equation.

---

### Step 2: Find the characteristic equation
Assume a solution of the form $\ ( y(x) = e^{rx} \ )$, where \( r \) is a constant. Substitute $\ ( y(x) = e^{rx} \ )$ into the differential equation:

$$
y''(x) - y(x) = 0
$$

First, compute \( y'(x) \) and \( y''(x) \):

$$
y'(x) = r e^{rx}, \quad y''(x) = r^2 e^{rx}.
$$

Substitute these into the equation:

$$
r^2 e^{rx} - e^{rx} = 0.
$$

Factor out \( e^{rx} \):

$$
e^{rx} (r^2 - 1) = 0.
$$

Since $\ ( e^{rx} \neq 0 \ )$, the characteristic equation is:

$$
r^2 - 1 = 0.
$$

---

### Step 3: Solve the characteristic equation
Solve the characteristic equation $\ ( r^2 - 1 = 0 \ )$:

$$
r^2 = 1 \implies r = \pm 1.
$$

This gives two roots:

$$
r = 1 \quad \text{and} \quad r = -1.
$$

---

### Step 4: Write the general solution
The general solution to the differential equation is a linear combination of the solutions corresponding to the roots of the characteristic equation. For \( r = 1 \) and \( r = -1 \), the general solution is:

$$
y(x) = C_1 e^{x} + C_2 e^{-x},
$$

where $\ ( C_1 \ ) and \ ( C_2 \ )$ are constants.

---

### Step 5: Apply the boundary conditions
We now use the boundary conditions to find the constants $\ ( C_1 \ ) and \ ( C_2 \ )$.

#### Boundary condition 1: \( y(0) = 2 \)
Substitute \( x = 0 \) and \( y(0) = 2 \) into the general solution:

$$
2 = C_1 e^{0} + C_2 e^{0}.
$$

Simplify:

$$
2 = C_1 + C_2.
$$

#### Boundary condition 2: \( y'(0) = 0 \)
First, compute $\ ( y'(x) \ )$ from the general solution:

$$
y(x) = C_1 e^{x} + C_2 e^{-x}.
$$

Differentiate with respect to \( x \):

$$
y'(x) = C_1 e^{x} - C_2 e^{-x}.
$$

Substitute \( x = 0 \) and $\ ( y'(0) = 0 \ )$:

$$
0 = C_1 e^{0} - C_2 e^{0}.
$$

Simplify:

$$
0 = C_1 - C_2 \implies C_1 = C_2.
$$

---

### Step 6: Solve for $\ ( C_1 \ ) and \ ( C_2 \ )$
From the first boundary condition, we have:

$$
2 = C_1 + C_2.
$$

Substitute $\ ( C_1 = C_2 \ )$:

$$
2 = C_1 + C_1 \implies 2 = 2C_1 \implies C_1 = 1.
$$

Since $\ ( C_1 = C_2 \ ), we also have \ ( C_2 = 1 \ )$.

---

### Step 7: Write the particular solution
Substitute $\ ( C_1 = 1 \ ) and \ ( C_2 = 1 \ )$ into the general solution:

$$
y(x) = e^{x} + e^{-x}.
$$

---

### Final Answer:
The solution to the differential equation \( y''(x) - y(x) = 0 \) with the boundary conditions \( y(0) = 2 \) and \( y'(0) = 0 \) is:

$$
y(x) = e^{x} + e^{-x}.
$$

---

### Verification:
To verify the solution, compute \( y'(x) \) and \( y''(x) \):

$$
y(x) = e^{x} + e^{-x},
$$

$$
y'(x) = e^{x} - e^{-x},
$$

$$
y''(x) = e^{x} + e^{-x}.
$$

Substitute into the differential equation:

$$
y''(x) - y(x) = (e^{x} + e^{-x}) - (e^{x} + e^{-x}) = 0.
$$

The solution satisfies the differential equation. Additionally:
- $\ ( y(0) = e^{0} + e^{0} = 2 \ )$,
- $\ ( y'(0) = e^{0} - e^{0} = 0 \ )$.

### Step 1: Rewrite the equation
The given differential equation is:

$$
\frac{d^2 y(x)}{dx^2} = -\omega^2 y(x).
$$

This can be rewritten as:

$$
y''(x) + \omega^2 y(x) = 0.
$$

This is a standard form of a **simple harmonic oscillator** equation.

---

### Step 2: Find the characteristic equation
Assume a solution of the form $\ ( y(x) = e^{rx} \ )$, where \( r \) is a constant. Substitute $\ ( y(x) = e^{rx} \ )$ into the differential equation:

$$
y''(x) + \omega^2 y(x) = 0.
$$

First, compute \( y'(x) \) and \( y''(x) \):

$$
y'(x) = r e^{rx}, \quad y''(x) = r^2 e^{rx}.
$$

Substitute these into the equation:

$$
r^2 e^{rx} + \omega^2 e^{rx} = 0.
$$

Factor out $\ ( e^{rx} \ )$:

$$
e^{rx} (r^2 + \omega^2) = 0.
$$

Since $\ ( e^{rx} \neq 0 \ )$, the characteristic equation is:

$$
r^2 + \omega^2 = 0.
$$

---

### Step 3: Solve the characteristic equation
Solve the characteristic equation $\ ( r^2 + \omega^2 = 0 \ )$:

$$
r^2 = -\omega^2 \implies r = \pm i\omega,
$$

where $\ ( i = \sqrt{-1} \ )$ is the imaginary unit. This gives two complex roots:

$$
r = i\omega \quad \text{and} \quad r = -i\omega.
$$

---

### Step 4: Write the general solution
For complex roots $\ ( r = \alpha \pm i\beta \ )$, the general solution to the differential equation is:

$$
y(x) = C_1 \cos(\beta x) + C_2 \sin(\beta x).
$$

In this case, $\ ( \alpha = 0 \ ) and \ ( \beta = \omega \ )$, so the general solution is:

$$
y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x),
$$

where $\ ( C_1 \ ) and \ ( C_2 \ )$ are constants determined by initial conditions.

---

### Step 5: Alternative form of the solution
The general solution can also be written in the form:

$$
