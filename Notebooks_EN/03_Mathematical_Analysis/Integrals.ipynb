Alright! Let’s break this down in a super simple way.  

Think of the symbol **∫** as a magic "sum-up" machine. It takes a function (a rule for numbers) and adds up all the tiny pieces over a certain range.  

Now:  

$$
\int 1 \,dx
$$

This means: "Add up a bunch of tiny pieces of 1 along the number line in terms of **x**."  

Since **1** is just a flat, never-changing number, the sum just grows like a staircase. The answer is:  

$$
x + C
$$

That **+ C** is because we can shift the whole staircase up or down, and it still works the same way.  

Answer is: **x + C**!

##2.Integral of  $\ ( x^2 + 2 \ )$  

### **Step 1: Understand the Question**
We need to find:

$$
\int (x^2 + 2) \,dx
$$

This means we are **adding up** all the tiny pieces of $\ ( x^2 + 2 \ )$.

---

### **Step 2: Break It Into Two Parts**
We can split this into two simpler problems:

$$
\int x^2 \,dx + \int 2 \,dx
$$

Now, let's solve **each part separately**.

---

### Step 3: Solve $\ ( \int x^2 \,dx \ )$
This follows the **Power Rule**, which says:

$$
\int x^n \,dx = \frac{x^{n+1}}{n+1}
$$

- Here, \( n = 2 \).
- Add **1** to the exponent: \( 2 + 1 = 3 \).
- Divide by **3**.

So,

$$
\int x^2 \,dx = \frac{x^3}{3}
$$

---

### Step 4: Solve $\ ( \int 2 \,dx \ )$
The number **2** is just a constant.  
When integrating a **constant**, just **multiply it by \( x \)**.

$$
\int 2 \,dx = 2x
$$

---

### **Step 5: Put It All Together**
Now we add both results:

$$
\frac{x^3}{3} + 2x + C
$$

  Answer:
$$
\frac{x^3}{3} + 2x + C
$$

Where **\( C \)** is the **constant of integration** (it appears in all indefinite integrals).

---

## 2. Integral of $\ ( 2\sin(x) \ )$  
### Step 1: Understand the Question**
We need to solve:

$$
\int 2\sin(x) \,dx
$$

---

### Step 2: Factor Out the 2
Since **2** is just a number, we can **move it outside**:

$$
2 \int \sin(x) \,dx
$$

---

### Step 3: Solve $\ ( \int \sin(x) \,dx \ )$
From known integrals:

$$
\int \sin(x) \,dx = -\cos(x)
$$

---

### **Step 4: Multiply by 2**
Now we multiply everything by **2**:

$$
2(-\cos(x)) = -2\cos(x)
$$

---

### **Final Answer:**
$$
-2\cos(x) + C
$$

---

## **3. Integral of \( 3x \)**
### **Step 1: Understand the Question**
We need to integrate:

$$
\int 3x \,dx
$$

---

### **Step 2: Use the Power Rule**
Since **\( x \) has an exponent of 1**, we apply the **Power Rule**:

$$
\int x^1 \,dx = \frac{x^{1+1}}{1+1}
$$

- Add **1** to the exponent: \( 1 + 1 = 2 \).
- Divide by **2**.

$$
\frac{x^2}{2}
$$

Now, multiply by **3**:

$$
\frac{3x^2}{2}
$$

---

### **Final Answer:**
$$
\frac{3x^2}{2} + C
$$

## 4. Integral of $\ ( \frac{1}{x^2} \ )$
### **Step 1: Rewrite the Expression**
Instead of writing $\ ( \frac{1}{x^2} \ )$, we write it as $\ ( x^{-2} \ )$:

$$
\int x^{-2} \,dx
$$

---

### **Step 2: Use the Power Rule**
Again, use:

$$
\int x^n \,dx = \frac{x^{n+1}}{n+1}
$$

- Here, \( n = -2 \).
- Add **1** to the exponent: \( -2 + 1 = -1 \).
- Divide by **-1**:

$$
\frac{x^{-1}}{-1} = -\frac{1}{x}
$$

---

### **Final Answer:**
$$
-\frac{1}{x} + C
$$

## **11 Integral:**
### Solve:
$$\int \frac{\cos(\ln x)}{x} \, dx$$

### **Step 1: Use Substitution**
Let:
$$ u = \ln x $$
Then:
$$ du = \frac{dx}{x} $$
Rewriting the integral:
$$ \int \cos u \, du $$

### **Step 2: Integrate**
$$ \int \cos u \, du = \sin u + C $$

### **Step 3: Substitute Back \( u = \ln x \)**
**Final Answer :**
$$ \sin(\ln x) + C $$

---

## **12 Integral:**
### Solve:
$$\int x \ln x \, dx$$

### **Step 1: Use Integration by Parts**
Formula:
$$\int u \, dv = uv - \int v \, du$$

Let:
- \( u = \ln x \Rightarrow du = \frac{dx}{x} \)
- \( dv = x \, dx \Rightarrow v = \frac{x^2}{2} \)

### **Step 2: Apply the Formula**
$$ \int x \ln x \, dx = \ln x \cdot \frac{x^2}{2} - \int \frac{x^2}{2} \cdot \frac{dx}{x} $$
Rewriting:
$$ = \frac{x^2}{2} \ln x - \int \frac{x}{2} dx $$

### **Step 3: Integrate**
$$\int \frac{x}{2} dx = \frac{1}{2} \int x \, dx$$
$$= \frac{1}{2} \times \frac{x^2}{2}$$
$$= \frac{x^2}{4}$$

**Final Answer :**
$$ \frac{x^2}{2} \ln x - \frac{x^2}{4} + C $$

---

## **13 Integral:**
### Solve:
$$\int x e^x \, dx$$

### **Step 1: Use Integration by Parts**
Let:
- \( u = x \Rightarrow du = dx \)
- \( dv = e^x \, dx \Rightarrow v = e^x \)

### **Step 2: Apply the Formula**
$$ \int x e^x \, dx = x e^x - \int e^x \, dx $$
Since:
$$ \int e^x \, dx = e^x $$

**Final Answer :**
$$ x e^x - e^x + C $$

# Calculating the Area of the Bounded Region

We need to find the **area** of the region bounded by the lines:
- \( x = 1 \)
- \( x = 2 \)
- \( y = 0 \) (the x-axis)
- The curve \( y = x^2 + 1 \)

---

## **Step 1: Set Up the Integral**

The area is given by the definite integral:
$$ A = \int_1^2 (x^2 + 1) \, dx $$

---

## Step 2: Compute the Integral

We split the integral into two parts:
$$ A = \int_1^2 x^2 \, dx + \int_1^2 1 \, dx $$

### Step 2.1: Solve $\ ( \int x^2 \, dx \ )$

Using the power rule:
$$ \int x^n dx = \frac{x^{n+1}}{n+1} $$

For $\ ( x^2 \ )$:
$$ \int x^2 dx = \frac{x^3}{3} $$

Evaluating from \( x = 1 \) to \( x = 2 \):
$$ \left( \frac{x^3}{3} \right) \Big|_1^2 = \frac{2^3}{3} - \frac{1^3}{3} = \frac{8}{3} - \frac{1}{3} = \frac{7}{3} $$

---

### Step 2.2: Solve $\ ( \int 1 \, dx \ )$

The integral of \( 1 \) is:
$$ \int 1 \, dx = x $$

Evaluating from \( x = 1 \) to \( x = 2 \):
$$ (x) \Big|_1^2 = 2 - 1 = 1 $$

---

## Step 3:Final Area**

Adding the results:
$$ A = \frac{7}{3} + 1 = \frac{7}{3} + \frac{3}{3} = \frac{10}{3} $$

The **area of the region is**:
$$ A = \frac{10}{3} \text{ square units.} $$
