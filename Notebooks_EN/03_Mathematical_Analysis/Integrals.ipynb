##2.Integral of  $\ ( x^2 + 2 \ )$  

### **Step 1: Understand the Question**
We need to find:

$$
\int (x^2 + 2) \,dx
$$

This means we are **adding up** all the tiny pieces of $\ ( x^2 + 2 \ )$.

---

### **Step 2: Break It Into Two Parts**
We can split this into two simpler problems:

$$
\int x^2 \,dx + \int 2 \,dx
$$

Now, let's solve **each part separately**.

---

### Step 3: Solve $\ ( \int x^2 \,dx \ )$
This follows the **Power Rule**, which says:

$$
\int x^n \,dx = \frac{x^{n+1}}{n+1}
$$

- Here, \( n = 2 \).
- Add **1** to the exponent: \( 2 + 1 = 3 \).
- Divide by **3**.

So,

$$
\int x^2 \,dx = \frac{x^3}{3}
$$

---

### Step 4: Solve $\ ( \int 2 \,dx \ )$
The number **2** is just a constant.  
When integrating a **constant**, just **multiply it by \( x \)**.

$$
\int 2 \,dx = 2x
$$

---

### **Step 5: Put It All Together**
Now we add both results:

$$
\frac{x^3}{3} + 2x + C
$$

  Answer:
$$
\frac{x^3}{3} + 2x + C
$$

Where **\( C \)** is the **constant of integration** (it appears in all indefinite integrals).

1. $\ ( f(x) = 2x + 1 \ )$
2. $\ ( g(x) = x^2 \ )$

These integrals help us find the **area under the curve** of each function from \( x = 0 \) to $\ ( x = \pi \ )$.

Think of it like pouring sand under the graphâ€™s curve and asking, **"How much sand is there?"**

---

## **1st Integral: $\ ( f(x) = 2x + 1 \ )$**
We need to compute:

$$
\int_0^\pi (2x + 1) \,dx
$$

### **Step 1: Break it into Parts**
Using the **sum rule**, we split the integral:

$$
\int (2x + 1) \,dx = \int 2x \,dx + \int 1 \,dx.
$$

This means we solve each part **separately**.

---

### **Step 2: Compute $\ ( \int 2x \,dx \ )$**
Using the **power rule**:

$$
\int x^n \,dx = \frac{x^{n+1}}{n+1}.
$$

For $\ ( 2x = 2x^1 \ )$:

1. Add **1** to the exponent:

   $$
   1 + 1 = 2.
   $$

2. Divide by the new exponent:

   $$
   \frac{x^2}{2}.
   $$

3. Multiply by 2:

   $$
   2 \times \frac{x^2}{2} = x^2.
   $$

So:

$$
\int 2x \,dx = x^2.
$$

---

### **Step 3: Compute $\ ( \int 1 \,dx \ )$**
The integral of **a constant** is just:

$$
\int 1 \,dx = x.
$$

---

### **Step 4: Combine Everything**
$$
\int (2x + 1) \,dx = x^2 + x.
$$

Since we are computing a **definite integral**, we evaluate at the limits:

$$
\left[ x^2 + x \right]_0^\pi
$$

---

### **Step 5: Plug in Upper Limit $( \ ( x = \pi \ ) )$**
$$
\pi^2 + \pi.
$$

### **Step 6: Plug in Lower Limit $( \ ( x = 0 \ ) )$**
$$
0^2 + 0 = 0.
$$

### **Step 7: Compute the Final Answer**
$$
(\pi^2 + \pi) - (0) = \pi^2 + \pi.
$$

So the **area under $\ ( f(x) = 2x + 1 \ ) from \ ( 0 \ ) to \ ( \pi \ )$ is:**

$$
\pi^2 + \pi.
$$

### **Step 5: Compute the Final Answer**
$$
\frac{\pi^3}{3} - 0 = \frac{\pi^3}{3}.

So the **area under $\ ( g(x) = x^2 \ ) from \ ( 0 \ ) to \ ( \pi \ ) $ is:**

# Calculating the Area of the Bounded Region

We need to find the **area** of the region bounded by the lines:
- \( x = 1 \)
- \( x = 2 \)
- \( y = 0 \) (the x-axis)
- The curve \( y = x^2 + 1 \)

---

## **Step 1: Set Up the Integral**

The area is given by the definite integral:
$$ A = \int_1^2 (x^2 + 1) \, dx $$

---

## Step 2: Compute the Integral

We split the integral into two parts:
$$ A = \int_1^2 x^2 \, dx + \int_1^2 1 \, dx $$

### Step 2.1: Solve $\ ( \int x^2 \, dx \ )$

Using the power rule:
$$ \int x^n dx = \frac{x^{n+1}}{n+1} $$

For $\ ( x^2 \ )$:
$$ \int x^2 dx = \frac{x^3}{3} $$

Evaluating from \( x = 1 \) to \( x = 2 \):
$$ \left( \frac{x^3}{3} \right) \Big|_1^2 = \frac{2^3}{3} - \frac{1^3}{3} = \frac{8}{3} - \frac{1}{3} = \frac{7}{3} $$

---

### Step 2.2: Solve $\ ( \int 1 \, dx \ )$

The integral of \( 1 \) is:
$$ \int 1 \, dx = x $$

Evaluating from \( x = 1 \) to \( x = 2 \):
$$ (x) \Big|_1^2 = 2 - 1 = 1 $$

---

## Step 3:Final Area

Adding the results:
$$ A = \frac{7}{3} + 1 = \frac{7}{3} + \frac{3}{3} = \frac{10}{3} $$

The **area of the region is**:
$$ A = \frac{10}{3} \text{ square units.} $$

# **Step-by-Step Calculation of the Area Under the Sine Curve**

We are finding the **area under the sine curve** from \( x = 0 \) to $\ ( x = \pi \ )$. This means we are computing the integral:

$$ P = \int_0^\pi \sin(x) \, dx $$

---

## **Step 1: Understanding the Integral**

An integral **finds the area under a curve**. Here, the curve is $\ ( \sin(x) \ )$, and we are looking at the interval $\ ( [0, \pi] \ )$.

Imagine the sine wave as a smooth hill. We want to find how much space is underneath it from \( x = 0 \) to $\ ( x = \pi \ )$.

---

## **Step 2: Finding the Integral of** $\ ( \sin(x) \ )$

To find the area, we need to compute:
$$ \int \sin(x) \, dx $$

The rule for integrating $\ ( \sin(x) \ )$ is:
$$ \int \sin(x) \, dx = -\cos(x) $$

This means the integral of $\ ( \sin(x) \ )$ is **negative cosine**.

---

## **Step 3: Evaluating from** \( 0 \) **to** $\ ( \pi \ )$

Now, we plug in the values **from 0 to \( \pi \)**:
$$ P = \left[ -\cos(x) \right]_0^\pi $$

This means:
$$ P = -\cos(\pi) + \cos(0) $$

We know from trigonometry:
- $ \ ( \cos(\pi) = -1 \ )$
- $\ ( \cos(0) = 1 \ )$

So we substitute:
$$ P = -(-1) + 1 $$

Since **negative of negative is positive**:
$$ P = 1 + 1 = 2 $$




# **Task 4: Finding the Area Under the Sine Curve from $\ (0\ ) to \ ( \pi\ )$**  

## **Step 1: What Does "Area Under the Curve" Mean?**  

Imagine you have a wavy line, like a smooth hill. This hill is the graph of the **sine function**, which we write as:

$$
y = \sin(x)
$$

You want to figure out **how much space** is under this curve, between \( x = 0 \) and $\ ( x = \pi \ )$.  

To do this, we are going to **chop** the space into lots of tiny, thin rectangles and add up their areas. When we take the limit where the rectangles become **infinitely thin**, this process is called **integration**.

## **Step 2: Writing the Integral**  
In math, the area under a curve from \( x = a \) to \( x = b \) is found using an **integral**:

$$
\int_{a}^{b} f(x) \,dx
$$

Here, our function is $\ ( f(x) = \sin(x) \ )$, and we are looking between \( x = 0 \) and $\ ( x = \pi \ )$, so our integral is:

$$
\int_{0}^{\pi} \sin(x) \,dx
$$

This means: **Find the total area under the sine wave from \( x = 0 \) to $\ ( \pi \ )$ by adding up infinitely many tiny slices of width \( dx \).**

## **Step 3: Finding the Integral of $\ ( \sin(x) \ )$**  

We need to figure out what function, when we take its **derivative**, gives us $\ ( \sin(x) \ )$.  

It turns out that:

$$
\frac{d}{dx} (-\cos(x)) = \sin(x)
$$

So, we say:

$$
\int \sin(x) \,dx = -\cos(x) + C
$$

(Here, the **C** is a "constant of integration," but since we are doing a definite integral with limits, we donâ€™t need it.)

## **Step 4: Evaluating the Integral from $\ (0\ ) to \ (\pi\ )$**  

Now, we **plug in** the limits:

$$
\left[-\cos(x)\right]_{0}^{\pi}
$$

This means:  
1. First, find $\ ( -\cos(x) \ ) when \ ( x = \pi \ )$:  
   $$
   -\cos(\pi) = -(-1) = 1
   $$
2. Then, find $\ ( -\cos(x) \ ) when \ ( x = 0 \ )$:  
   $$
   -\cos(0) = -1
   $$
3. Now, subtract the two values:

$$
1 - (-1) = 1 + 1 = 2
$$

## **Final Answer for Task 4:**  
The area under the sine curve from $\ ( 0 \ ) to \ ( \pi \ )$ is **2 square units**!

# Task 5: Finding the Length of the Sine Curve from $\ (0\ ) to \ (\pi\ )$  

Now, instead of the area, we want to **measure how long the sine curve is** from $\ ( x = 0 \ ) to \ ( x = \pi \ )$.  

## **Step 1: What Does "Length of a Curve" Mean?**  

Imagine you are **walking** along the sine wave from $\ ( x = 0 \ ) to \ ( x = \pi \ )$. How far would you walk?  

This is **not just the difference in x-values** (which is $\ ( \pi - 0 = \pi \ )$ ). The curve is **bumpy**, so you actually walk a longer distance than just going straight across!  

To find this length, we need **calculus**.

## **Step 2: The Formula for Arc Length**  

The **length of a curve** from \( x = a \) to \( x = b \) is given by:

$$
L = \int_{a}^{b} \sqrt{1 + (f'(x))^2} \,dx
$$

This formula comes from the **Pythagorean theorem**, because every tiny step along the curve is a little slanted line, forming a tiny right triangle.

## Step 3: Finding $\ ( f'(x) \ )$    

Our function is:

$$
f(x) = \sin(x)
$$

Now, let's find its **derivative**:

$$
f'(x) = \cos(x)
$$

Now, square it:

$$
(f'(x))^2 = \cos^2(x)
$$

## **Step 4: Plugging into the Formula**  

Substituting this into the formula:

$$
L = \int_{0}^{\pi} \sqrt{1 + \cos^2(x)} \,dx
$$

This means we are adding up **tiny diagonal segments** of the sine wave from $\
 ( 0 \ ) to \ ( \pi \ )$.  

## **Step 5: Solving the Integral**  

The integral:

$$
\int_{0}^{\pi} \sqrt{1 + \cos^2(x)} \,dx
$$

We **approximate** it using numerical methods.

Using a calculator or software, the result is:

$$
L \approx 3.82
$$

## **Final Answer for Task 5:**  
The length of the sine curve from $\ ( 0 \ ) to \ ( \pi \ )$ is about **3.82 units**!

## Step 1: Understanding the Problem
We have a moving particle, and its position at any time \( t \) is given by:

$$
x(t) = 3t^2 - 6t + 1
$$

This function tells us where the particle is at any moment in time.

But we don't just want to know where it isâ€”we want to know how far it **actually traveled** between \( t = 0 \) and \( t = 2 \).

---

## Step 2: Find How Fast the Particle Moves
To find how fast the particle moves, we need to take the derivative of \( x(t) \). That just means we find how position changes over time.

The derivative of \( x(t) \) is:

$$
v(t) = \frac{d}{dt} (3t^2 - 6t + 1) = 6t - 6
$$

This is called **velocity**. It tells us whether the particle is moving forward or backward.

---

## Step 3: Find When the Particle Stops
If we want to know when the particle **changes direction**, we set \( v(t) = 0 \) (because at that moment, it's neither moving forward nor backward):

$$
6t - 6 = 0
$$

Solving for \( t \):

$$
t = 1
$$

So at \( t = 1 \), the particle **stops and turns around**.

---

## Step 4: Break the Problem into Two Parts
Since the particle changes direction at \( t = 1 \), we **canâ€™t** just find the difference between start and end points. Instead, we measure how much it moves:
- **From \( t = 0 \) to \( t = 1 \)** (Moving **backward**)
- **From \( t = 1 \) to \( t = 2 \)** (Moving **forward**)

---

## Step 5: Compute Distance in Each Part
Since velocity is **negative** in the first part (going backward), we take its absolute value (so we count all movement as positive).

### **First Part: From \( t = 0 \) to \( t = 1 \)**  
We integrate \( 6 - 6t \) (the absolute value of velocity):

$$
\int_0^1 (6 - 6t) \,dt
$$

This gives us **3 units**.

### **Second Part: From \( t = 1 \) to \( t = 2 \)**  
We integrate \( 6t - 6 \) (since velocity is now positive):

$$
\int_1^2 (6t - 6) \,dt
$$

This also gives us **3 units**.

---

## Step 6: Add Both Parts
Now, we add both distances together:

$$
3 + 3 = 6
$$

So, the total distance the particle traveled is **6 units**!
